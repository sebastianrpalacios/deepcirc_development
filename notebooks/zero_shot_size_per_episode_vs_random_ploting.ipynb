```python
# Trained vs Random(masked + backtrack)  best_energy_so_far vs episode
# Assumes outputs from the latest run script:
#  - trained:  <search_root>/trained_masked/trained_episode_metrics.csv
#  - random:   <search_root>/random_masked/random_episode_metrics.csv
#
# Episode 0 comes from trajectories/ep_0000.json (steps[0].energy) and is
# applied to BOTH curves (prepend/overwrite ep0).
#
# Formatted in the same GraphPad-style as your unrelated example.

import json
import math
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.ticker as mticker

# ---------------- GraphPad-style Matplotlib settings ----------------
mpl.rcParams.update({
    "font.family": "sans-serif",
    "font.sans-serif": ["DejaVu Sans"],
    "font.size": 6,
    "axes.labelsize": 6,
    "axes.titlesize": 6,
    "axes.linewidth": 0.5,
    "lines.linewidth": 0.5,
    "xtick.labelsize": 6,
    "ytick.labelsize": 6,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 5,
    "ytick.major.size": 5,
    "legend.fontsize": 6,
    "legend.frameon": False,
    "savefig.dpi": 300,
    "figure.dpi": 100,
    "figure.figsize": (8, 4),
    "axes.grid": False,
    "pdf.fonttype": 42,
    "ps.fonttype": 42,
    "pdf.use14corefonts": False,
    "svg.fonttype": "none",
})
pt_to_in = lambda pt: pt / 72

# ---------- spreadsheet once ----------
sheet_path = Path("/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/scripts/runs/Fig. 2 Sizes of random 4 input circuits plus hex verification and validation (trained from scratch and automated analysis for zero-shot comparison).csv")

def _norm_hex(x):
    s = str(x).strip().lower()
    if not s.startswith("0x"):
        s = "0x" + s
    return "0x" + s[2:].upper()

sizes_df = pd.read_csv(sheet_path, dtype={"circuit_hex": str})
sizes_df["circuit_hex_norm"] = sizes_df["circuit_hex"].map(_norm_hex)
sizes_df["smallest_ML_size"] = pd.to_numeric(sizes_df["smallest_ML_size"], errors="coerce")
size_lookup = dict(zip(sizes_df["circuit_hex_norm"], sizes_df["smallest_ML_size"]))



# ---------- colors & labels ----------
color_map = {
    "trained":  "#6abefe",  # DeepCirc (masked)
    "random":   "#b8bbbd",  # Random masked + backtrack
}
label_map = {
    "trained":  "DeepCirc",
    "random":   "Random (masked + backtrack)",
}

# ---------- helpers ----------
def find_paths(search_root):
    """Return (trained_csv, random_csv) if present; else None for missing."""
    trained_csv = None
    random_csv = None

    candidates_trained = [
        search_root / "trained_masked" / "trained_episode_metrics.csv",
        search_root / "trained_masked" / "trajectories" / "trained_episode_metrics.csv",  # fallback
    ]
    for c in candidates_trained:
        if c.exists():
            trained_csv = c
            break

    candidates_random = [
        search_root / "random_masked" / "random_episode_metrics.csv",
        search_root / "random_masked" / "trajectories" / "random_episode_metrics.csv",   # fallback
    ]
    for c in candidates_random:
        if c.exists():
            random_csv = c
            break

    return trained_csv, random_csv

def find_initial_energy(search_root):
    """
    Read steps[0].energy from ep_0000.json.
    Prefer trained_masked trajectory; if not found, try random_masked.
    """
    patterns = [
        search_root / "trained_masked" / "trajectories" / "ep_0000.json",
        search_root / "random_masked" / "trajectories" / "ep_0000.json",
    ]
    for p in patterns:
        if p.exists():
            try:
                with open(p, "r") as f:
                    data = json.load(f)
                return float(data["steps"][0]["energy"])
            except Exception as e:
                print(f"[WARN] Failed reading initial energy from {p}: {e}")
                return None
    return None

def read_episode_curve(csv_path, initial_energy):
    """
    Read episode-based CSV with columns: episode, best_energy_so_far.
    Returns x (episodes) and y (best_energy_so_far), injecting/overwriting ep0 with initial_energy if provided.
    """
    d = pd.read_csv(csv_path)
    if not {"episode", "best_energy_so_far"}.issubset(d.columns):
        raise ValueError(f"Missing required columns in {csv_path}")
    d = d.sort_values("episode")
    x = d["episode"].to_numpy()
    y = pd.to_numeric(d["best_energy_so_far"], errors="coerce").to_numpy(dtype=float)

    if initial_energy is not None:
        if (x == 0).any():
            y[x == 0] = initial_energy
        else:
            x = np.insert(x, 0, 0)
            y = np.insert(y, 0, initial_energy)
    return x, y

# ------------ GraphPad-style tick spacing ------------
BASE_TICK = 25 # episodes between major ticks

# ---------- main plotting loop ----------
for hex_id in hex_ids:
    search_root = Path(

        f"/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/manuscript/zero-shot/against_random/GAT_MLP_with_scalars/initial_state_sampling_factor_0/4_inputs_{hex_id}_samplingfactor_0_4000nn/seed_1"
    )

    trained_csv, random_csv = find_paths(search_root)
    initial_energy = find_initial_energy(search_root)  # ep0 from JSON for BOTH curves

    # --- figure size to match example (130pt  110pt) ---
    fig, ax = plt.subplots(figsize=(pt_to_in(130), pt_to_in(110)), dpi=150)
    loaded_any = False
    max_ep = 0

    # ---- trained curve ----
    if trained_csv and trained_csv.exists():
        try:
            x, y = read_episode_curve(trained_csv, initial_energy)
            if len(x) > 0:
                max_ep = max(max_ep, int(np.max(x)))
            ax.plot(x, y, label=label_map["trained"], color=color_map["trained"], linestyle='-', linewidth = 2)
            loaded_any = True
        except Exception as e:
            print(f"[{hex_id}] Failed to read trained CSV {trained_csv}: {e}")
    else:
        print(f"[{hex_id}] Missing trained metrics CSV: {trained_csv}")

    # ---- random curve ----
    if random_csv and random_csv.exists():
        try:
            xr, yr = read_episode_curve(random_csv, initial_energy)
            if len(xr) > 0:
                max_ep = max(max_ep, int(np.max(xr)))
            ax.plot(xr, yr, label=label_map["random"], color=color_map["random"], linestyle='-', linewidth = 2)
            loaded_any = True
        except Exception as e:
            print(f"[{hex_id}] Failed to read random CSV {random_csv}: {e}")
    else:
        print(f"[{hex_id}] Missing random metrics CSV: {random_csv}")

    # ---- dashed ML baseline ----
    ml_size = size_lookup.get(_norm_hex(hex_id))
    if pd.notna(ml_size):
        ax.axhline(
            y=float(ml_size),
            linestyle="--",
            linewidth=0.5,
            color="#444444",
            label="DeepCirc trained from scratch on single design",
        )

    if loaded_any:
        # Aesthetics like the example
        for spine in ax.spines.values():
            spine.set_linewidth(0.5)

        right_limit = int(np.ceil(max_ep / BASE_TICK) * BASE_TICK) if max_ep > 0 else BASE_TICK
        ax.xaxis.set_major_locator(mticker.MultipleLocator(BASE_TICK))
        #ax.set_xlim(0, right_limit)
        ax.set_xlim(-10, right_limit)
        ax.margins(x=0.02)
        ax.minorticks_off()
        ax.tick_params(axis='both', which='both', direction='out', length=2, width=0.5)

        ax.set_xlabel("Episode")
        ax.set_ylabel("Number of repressors")
        title_suffix = " (ep0 = initial from JSON)" if initial_energy is not None else ""
        ax.set_title(f"Circuit {hex_id}{title_suffix}")

        ax.legend(loc="center left", bbox_to_anchor=(1.02, 0.5), borderaxespad=0)
        plt.tight_layout(rect=[0, 0, 0.85, 1])
        plt.show()
    else:
        print(f"[{hex_id}] No curves could be plotted (missing/invalid CSVs).")
```

```python
# ---------- helpers for stats (refactored: plotting vs saving/showing) ----------
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from typing import Dict, List, Tuple, Optional

# ====== you already have these in scope ======
# sheet_path: Path or str
# hex_ids: List[str]

def _norm_hex(x: str) -> str:
    s = str(x).strip().lower()
    if not s.startswith("0x"):
        s = "0x" + s
    return "0x" + s[2:].upper()

def _load_size_lookup(sheet_path) -> Dict[str, float]:
    sizes_df = pd.read_csv(sheet_path, dtype={"circuit_hex": str})
    sizes_df["circuit_hex_norm"] = sizes_df["circuit_hex"].map(_norm_hex)
    sizes_df["smallest_ML_size"] = pd.to_numeric(sizes_df["smallest_ML_size"], errors="coerce")
    return dict(zip(sizes_df["circuit_hex_norm"], sizes_df["smallest_ML_size"]))

size_lookup = _load_size_lookup(sheet_path)

METHODS = ["trained", "trained_sampling0", "trained_no_registry", "trained_unmasked", "random"]
LABELS = {
    "trained":             "DeepCirc",
    "trained_sampling0":   "No weighted sampling",
    "trained_no_registry": "No registry",
    "trained_unmasked":    "No masking",
    "random":              "Random (masked)",
}
patterns = {
    "random": [
        "**/random_masked/trajectories/*episode_metrics.csv",
        "**/random_masked/*random_episode_metrics.csv",
        "**/*random*episode_metrics.csv",
    ],
    "trained": [
        "**/trained_masked/trajectories/*episode_metrics.csv",
        "**/trained_masked/*trained_episode_metrics.csv",
        "**/*trained_episode_metrics.csv",
    ],
    "trained_sampling0": [
        "**/trained_sampling0_masked/trajectories/*episode_metrics.csv",
        "**/trained_sampling0_masked/*trained_sampling0_episode_metrics.csv",
        "**/*trained_sampling0*episode_metrics.csv",
    ],
    "trained_no_registry": [
        "**/trained_masked_no_registry/trajectories/*episode_metrics.csv",
        "**/trained_masked_no_registry/*trained_no_registry_episode_metrics.csv",
        "**/*trained_no_registry*episode_metrics.csv",
    ],
    "trained_unmasked": [
        "**/trained_unmasked/trajectories/*episode_metrics.csv",
        "**/trained_unmasked/*trained_unmasked_episode_metrics.csv",
        "**/*trained_unmasked*episode_metrics.csv",
    ],
}

def find_csv(search_root: Path, method: str) -> Optional[Path]:
    root = Path(search_root)
    for pat in patterns[method]:
        matches = sorted(set(p.as_posix() for p in root.glob(pat)))
        if matches:
            return Path(matches[0])
    if method == "random":
        ex = root / "random_episode_metrics.csv"
        if ex.exists():
            return ex
    return None

def get_final_size(csv_path: Path) -> Optional[float]:
    try:
        df = pd.read_csv(csv_path)
        if "episode" in df.columns:
            df = df.sort_values("episode", kind="mergesort")
        if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
            return float(pd.to_numeric(df["best_energy_so_far"], errors="coerce").dropna().iloc[-1])
        if "best_energy_in_episode" in df.columns and df["best_energy_in_episode"].notna().any():
            s = pd.to_numeric(df["best_energy_in_episode"], errors="coerce")
            cm = s.cummin()
            if cm.notna().any():
                return float(cm.dropna().iloc[-1])
    except Exception as e:
        print(f"[WARN] Failed to read {csv_path}: {e}")
    return None

def _root_for_hex(hex_id: str) -> Path:
    return Path(f"/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/manuscript/zero-shot/against_random/GAT_MLP_with_scalars/initial_state_sampling_factor_0/4_inputs_{hex_id}_samplingfactor_0_4000nn/seed_1")

def compare_method_to_ml(method: str, hex_ids: List[str]) -> Tuple[Dict[str, int], int, List[str]]:
    outcomes = {"Win": 0, "Loss": 0, "Tie": 0}
    n = 0
    missing = []
    for hex_id in hex_ids:
        root = _root_for_hex(hex_id)
        csv_path = find_csv(root, method)
        final_size = get_final_size(csv_path) if csv_path else None
        ml_val = size_lookup.get(_norm_hex(hex_id))
        if csv_path is None or final_size is None or ml_val is None or np.isnan(ml_val):
            missing.append(hex_id)
            continue
        n += 1
        if final_size < ml_val:
            outcomes["Win"] += 1
        elif final_size > ml_val:
            outcomes["Loss"] += 1
        else:
            outcomes["Tie"] += 1
    return outcomes, n, missing

def to_percentages(counts: Dict[str, int], denom: int) -> List[float]:
    if denom == 0:
        return [0.0, 0.0, 0.0]
    return [100.0 * counts["Win"]/denom, 100.0 * counts["Loss"]/denom, 100.0 * counts["Tie"]/denom]

# ======== PLOTTING: creation only (no save/show here) ========
def make_bar_figure(categories: List[str], values: List[float], title: str) -> plt.Figure:
    fig, ax = plt.subplots(figsize=(4, 3))
    bars = ax.bar(categories, values)
    ax.set_ylim(0, 100)
    ax.set_ylabel("Percentage of circuits")
    ax.set_title(title)
    for rect, v in zip(bars, values):
        ax.text(rect.get_x() + rect.get_width()/2, rect.get_height() + 1,
                f"{v:.1f}%", ha="center", va="bottom", fontsize=9)
    fig.tight_layout()
    return fig

def make_multi_method_figure(results_df: pd.DataFrame, title: str) -> plt.Figure:
    df = results_df.set_index("method").loc[METHODS].reset_index()
    cats = ["Win_%", "Loss_%", "Tie_%"]
    x = np.arange(len(df))
    width = 0.25
    fig, ax = plt.subplots(figsize=(6.5, 3.6))
    bars = []
    for i, cat in enumerate(cats):
        bars.append(ax.bar(x + (i-1)*width, df[cat].values, width, label=cat.replace("_", " ")))
    ax.set_ylabel("Percentage of circuits")
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels([LABELS[m] for m in df["method"]], rotation=15, ha="right")
    for grp in bars:
        for rect in grp:
            h = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2, h + 0.8, f"{h:.1f}%", ha="center", va="bottom", fontsize=8)
    ax.set_ylim(0, 100)
    ax.legend(frameon=False, ncols=3, loc="upper center", bbox_to_anchor=(0.5, 1.18))
    fig.tight_layout()
    return fig

# ======== SAVE/SHOW utilities (optional) ========
def save_fig(fig: plt.Figure, path: Path, dpi: int = 200):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    fig.savefig(path, dpi=dpi, bbox_inches="tight")

def show_fig(fig: plt.Figure):
    plt.show()

def close_fig(fig: plt.Figure):
    plt.close(fig)

# ======== Orchestrator with optional plotting/saving/showing ========
def summarize_and_plot(
    hex_ids: List[str],
    make_plots: bool = True,
    show_plots: bool = False,
    save_summary_csv: Optional[Path] = None,
    save_per_method_dir: Optional[Path] = None,
    save_multi_plot_path: Optional[Path] = None,
):
    """
    Returns (summary_df, per_method_figs, multi_method_fig)
      - per_method_figs: dict[method] -> Figure (or {})
      - multi_method_fig: Figure or None

    Control behavior with flags:
      make_plots: create figures (no saving/showing unless you ask)
      show_plots: if True, calls show_fig(fig) for created figures
      save_per_method_dir: if provided, saves per-method PNGs into this dir
      save_multi_plot_path: if provided, saves the multi-method PNG at this path
    """
    cats = ["Win", "Loss", "Tie"]
    rows = []
    per_method_figs: Dict[str, plt.Figure] = {}
    multi_method_fig: Optional[plt.Figure] = None

    for m in METHODS:
        outcomes_m, n_m, missing_m = compare_method_to_ml(m, hex_ids)
        pct_m = to_percentages(outcomes_m, n_m)
        print(f"{LABELS[m]} vs smallest_ML_size  counts: {outcomes_m} out of N={n_m} (missing {len(missing_m)})")

        if make_plots:
            title = f"{LABELS[m]} vs trained-from-scratch baseline (N={n_m})"
            fig = make_bar_figure(cats, pct_m, title)
            per_method_figs[m] = fig
            if save_per_method_dir is not None:
                save_fig(fig, Path(save_per_method_dir) / f"{m}_vs_baseline_N{n_m}.png")
            if show_plots:
                show_fig(fig)

        rows.append({
            "method": m,
            "label": LABELS[m],
            "N_used": n_m,
            "Win": outcomes_m["Win"],
            "Loss": outcomes_m["Loss"],
            "Tie": outcomes_m["Tie"],
            "Win_%": pct_m[0],
            "Loss_%": pct_m[1],
            "Tie_%": pct_m[2],
            "N_missing": len(missing_m),
        })

    summary = pd.DataFrame(rows, columns=["method","label","N_used","Win","Loss","Tie","Win_%","Loss_%","Tie_%","N_missing"])

    if make_plots:
        multi_method_fig = make_multi_method_figure(summary, "DeepCirc vs baseline (all methods)")
        if save_multi_plot_path is not None:
            save_fig(multi_method_fig, save_multi_plot_path)
        if show_plots:
            show_fig(multi_method_fig)

    if save_summary_csv is not None:
        Path(save_summary_csv).parent.mkdir(parents=True, exist_ok=True)
        summary.to_csv(save_summary_csv, index=False)

    return summary, per_method_figs, multi_method_fig

# ======================== example usage ========================
summary_df, figs, multi_fig = summarize_and_plot(
     hex_ids,
     make_plots=True,           # set False to skip creating any figures
     show_plots=True,          # set True to display figures interactively
 )
# # If you prefer to save some figures later:
# # save_fig(figs["trained"], Path("./plots/custom_trained.png"))
# # close_fig(figs["trained"])
```

```python
# ==== Trained vs Random: single comparison figure ====
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from typing import Optional, Tuple, Dict, List

# expects you already have:
# - compare_method_to_ml(method: str, hex_ids: List[str]) -> (counts, n_used, missing)
# - to_percentages(counts, denom) -> [win%, loss%, tie%]
# - LABELS dict with keys "trained" and "random"

def make_trained_vs_random_figure(
    hex_ids: List[str],
    title: str = "DeepCirc vs Random (Win/Loss/Tie)",
) -> Tuple[plt.Figure, Dict[str, Dict[str, float]]]:
    """
    Returns:
      fig: matplotlib Figure (not shown/saved)
      stats: {
        'trained': {'Win_%':..., 'Loss_%':..., 'Tie_%':..., 'N_used': int},
        'random':  {'Win_%':..., 'Loss_%':..., 'Tie_%':..., 'N_used': int},
      }
    """
    cats = ["Win", "Loss", "Tie"]
    m1, m2 = "trained", "random"

    # gather stats
    out1, n1, _ = compare_method_to_ml(m1, hex_ids)
    out2, n2, _ = compare_method_to_ml(m2, hex_ids)
    p1 = to_percentages(out1, n1)
    p2 = to_percentages(out2, n2)

    stats = {
        m1: {"Win_%": p1[0], "Loss_%": p1[1], "Tie_%": p1[2], "N_used": n1},
        m2: {"Win_%": p2[0], "Loss_%": p2[1], "Tie_%": p2[2], "N_used": n2},
    }

    # figure: 3 groups (Win/Loss/Tie), 2 bars per group (trained vs random)
    x = np.arange(len(cats))
    width = 0.35

    fig, ax = plt.subplots(figsize=(5.4, 3.4))
    bars1 = ax.bar(x - width/2, p1, width, label=f"{LABELS[m1]} (N={n1})")
    bars2 = ax.bar(x + width/2, p2, width, label=f"{LABELS[m2]} (N={n2})")

    ax.set_ylim(0, 100)
    ax.set_ylabel("Percentage of circuits")
    ax.set_title(title)
    ax.set_xticks(x, cats)
    ax.legend(frameon=False, loc="upper center", ncols=2, bbox_to_anchor=(0.5, 1.18))

    # annotate bar tops
    for grp in (bars1, bars2):
        for rect in grp:
            h = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2, h + 0.8, f"{h:.1f}%",
                    ha="center", va="bottom", fontsize=8)

    fig.tight_layout()
    return fig, stats

# ===== optional save/show helpers (keep plotting separate and optional) =====
def save_fig(fig: plt.Figure, path: Path, dpi: int = 200):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    fig.savefig(path, dpi=dpi, bbox_inches="tight")

def show_fig(fig: plt.Figure):
    plt.show()

def close_fig(fig: plt.Figure):
    plt.close(fig)

# ======================== example usage ========================
fig, stats = make_trained_vs_random_figure(hex_ids, title="DeepCirc vs Random (Win/Loss/Tie)")
print(stats)  # view percentages and N_used
#save_fig(fig, Path("./plots/trained_vs_random.png"))   # optional
show_fig(fig)
close_fig(fig)  # free memory if saving many figures
```

```python
# ==== Pairwise comparison: trained vs random (no baseline) ====
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from typing import List, Dict, Tuple, Optional

# Expects these to already exist in your session:
# - _root_for_hex(hex_id: str) -> Path
# - find_csv(search_root: Path, method: str) -> Optional[Path]
# - get_final_size(csv_path: Path) -> Optional[float]
# - LABELS with keys "trained" and "random" (e.g., {"trained": "DeepCirc", "random": "Random (masked)"})


def compare_trained_vs_random(hex_ids: List[str]) -> Tuple[Dict[str, int], int, List[str]]:
    """
    Pairwise per-circuit comparison:
      - 'Trained better'  if final_size(trained) < final_size(random)
      - 'Random better'   if final_size(trained) > final_size(random)
      - 'Tie'             if equal

    Returns: (counts, n_used, missing_hexes)
    """
    counts = {"Trained better": 0, "Random better": 0, "Tie": 0}
    n_used = 0
    missing = []

    for h in hex_ids:
        root = _root_for_hex(h)

        p_tr = find_csv(root, "trained")
        p_ra = find_csv(root, "random")

        s_tr = get_final_size(p_tr) if p_tr else None
        s_ra = get_final_size(p_ra) if p_ra else None

        if s_tr is None or s_ra is None:
            missing.append(h)
            continue

        n_used += 1
        if s_tr < s_ra:
            counts["Trained better"] += 1
        elif s_tr > s_ra:
            counts["Random better"] += 1
        else:
            counts["Tie"] += 1

    return counts, n_used, missing


def to_percentages(counts: Dict[str, int], denom: int) -> List[float]:
    if denom == 0:
        return [0.0, 0.0, 0.0]
    return [
        100.0 * counts["Trained better"] / denom,
        100.0 * counts["Random better"] / denom,
        100.0 * counts["Tie"] / denom,
    ]


# -------- Plot creation ONLY (no saving/showing) --------
def make_pairwise_trained_random_figure(
    counts: Dict[str, int],
    n_used: int,
    title: Optional[str] = None,
) -> plt.Figure:
    """
    Build a bar chart for Trained vs Random (pairwise). Returns a Figure.
    """
    cats = ["Trained better", "Random better", "Tie"]
    pct = to_percentages(counts, n_used)

    if title is None:
        title = f"{LABELS['trained']} vs {LABELS['random']} (pairwise, N={n_used})"

    fig, ax = plt.subplots(figsize=(5.4, 3.4))
    bars = ax.bar(cats, pct)
    ax.set_ylim(0, 100)
    ax.set_ylabel("Percentage of circuits")
    ax.set_title(title)

    # annotate tops
    for rect, v in zip(bars, pct):
        ax.text(rect.get_x() + rect.get_width()/2, rect.get_height() + 0.8,
                f"{v:.1f}%", ha="center", va="bottom", fontsize=9)

    fig.tight_layout()
    return fig


# -------- Optional save/show helpers --------
def save_fig(fig: plt.Figure, path: Path, dpi: int = 200):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    fig.savefig(path, dpi=dpi, bbox_inches="tight")

def show_fig(fig: plt.Figure):
    plt.show()

def close_fig(fig: plt.Figure):
    plt.close(fig)


# ================= Example usage =================
counts, n_used, missing = compare_trained_vs_random(hex_ids)
print("Pairwise counts:", counts, "N_used:", n_used, "missing:", len(missing))
fig = make_pairwise_trained_random_figure(counts, n_used)
# # Optional:
# # save_fig(fig, Path("./plots/pairwise_trained_vs_random.png"))
show_fig(fig)
close_fig(fig)
```
