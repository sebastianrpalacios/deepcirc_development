```python
# SA (masked + backtrack)  best_energy_so_far vs environment steps
# Assumes outputs from the SA script:
#  - sa per-step metrics: <search_root>/sa_masked/sa_step_metrics.csv
#  - trajectory JSON (for initial energy): <search_root>/sa_masked/trajectories/ep_0000.json
#
# NOTE: If 'best_energy_so_far' is missing, we compute it as a cumulative min over 'energy'.

import json
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.ticker as mticker

# ---------------- GraphPad-style Matplotlib settings ----------------
mpl.rcParams.update({
    "font.family": "sans-serif",
    "font.sans-serif": ["DejaVu Sans"],
    "font.size": 6,
    "axes.labelsize": 6,
    "axes.titlesize": 6,
    "axes.linewidth": 0.5,
    "lines.linewidth": 0.5,
    "xtick.labelsize": 6,
    "ytick.labelsize": 6,
    "xtick.direction": "in",
    "ytick.direction": "in",
    "xtick.major.size": 5,
    "ytick.major.size": 5,
    "legend.fontsize": 6,
    "legend.frameon": False,
    "savefig.dpi": 300,
    "figure.dpi": 100,
    "figure.figsize": (8, 4),
    "axes.grid": False,
    "pdf.fonttype": 42,
    "ps.fonttype": 42,
    "pdf.use14corefonts": False,
    "svg.fonttype": "none",
})
pt_to_in = lambda pt: pt / 72

# ---------- spreadsheet once ----------
sheet_path = Path("/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/scripts/runs/Fig. 2 Sizes of random 4 input circuits plus hex verification and validation (trained from scratch and automated analysis for zero-shot comparison).csv")

def _norm_hex(x):
    s = str(x).strip().lower()
    if not s.startswith("0x"):
        s = "0x" + s
    return "0x" + s[2:].upper()

sizes_df = pd.read_csv(sheet_path, dtype={"circuit_hex": str})
sizes_df["circuit_hex_norm"] = sizes_df["circuit_hex"].map(_norm_hex)
sizes_df["smallest_ML_size"] = pd.to_numeric(sizes_df["smallest_ML_size"], errors="coerce")
size_lookup = dict(zip(sizes_df["circuit_hex_norm"], sizes_df["smallest_ML_size"]))

# ---------- colors & labels ----------
SA_COLOR = "#b8bbbd"
SA_LABEL = "Simulated Annealing (masked + backtrack)"

# ---------- helpers ----------
def find_sa_paths(search_root):
    """Return (sa_step_csv, ep0_json) if present; else None for missing."""
    sa_csv = search_root / "sa_masked" / "sa_step_metrics.csv"
    ep0 = search_root / "sa_masked" / "trajectories" / "ep_0000.json"
    return (sa_csv if sa_csv.exists() else None,
            ep0 if ep0.exists() else None)

def read_initial_energy(ep0_json):
    if ep0_json is None:
        return None
    try:
        with ep0_json.open("r") as f:
            data = json.load(f)
        return float(data["steps"][0]["energy"])
    except Exception as e:
        print(f"[WARN] Failed reading initial energy from {ep0_json}: {e}")
        return None

def read_sa_step_curve(csv_path, initial_energy):
    """
    Read per-step CSV with columns: step, energy, (optional) best_energy_so_far.
    Returns arrays (steps, best_so_far). If initial_energy is provided, we inject it at the
    smallest step (usually 0) and recompute best-so-far to ensure consistency.
    """
    d = pd.read_csv(csv_path)
    if "step" not in d.columns:
        raise ValueError(f"Missing 'step' column in {csv_path}")
    d = d.sort_values("step").reset_index(drop=True)

    # Ensure numeric
    d["step"] = pd.to_numeric(d["step"], errors="coerce")
    if "best_energy_so_far" in d.columns:
        d["best_energy_so_far"] = pd.to_numeric(d["best_energy_so_far"], errors="coerce")
    if "energy" in d.columns:
        d["energy"] = pd.to_numeric(d["energy"], errors="coerce")

    # Compute best-so-far if needed
    if "best_energy_so_far" not in d.columns or d["best_energy_so_far"].isna().all():
        if "energy" not in d.columns:
            raise ValueError(f"'best_energy_so_far' missing and 'energy' not found in {csv_path}")
        d["best_energy_so_far"] = d["energy"].cummin()

    # Inject/overwrite initial energy at the first step (if provided)
    if initial_energy is not None and len(d) > 0:
        first_idx = 0
        # overwrite best at first step to reflect initial
        d.loc[first_idx, "best_energy_so_far"] = min(
            float(d.loc[first_idx, "best_energy_so_far"]), float(initial_energy)
        )
        # recompute cumulative min from the first row forward
        d["best_energy_so_far"] = d["best_energy_so_far"].cummin()

    # Drop any rows with NaN step or best
    d = d.dropna(subset=["step", "best_energy_so_far"])
    return d["step"].to_numpy(dtype=int), d["best_energy_so_far"].to_numpy(dtype=float)

# ------------ tick spacing ------------
BASE_TICK_STEPS = 4000  # steps between major x ticks

# ---------- main plotting loop ----------
# Expect `hex_ids` to be defined upstream (list of circuit hex IDs to process)
for hex_id in hex_ids:
    search_root = Path(
        f"/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/manuscript/logic_synthesis_SA/11400_steps/4_inputs/no_fw_preference/{hex_id}/seed_1"
    )

    sa_csv, ep0_json = find_sa_paths(search_root)
    initial_energy = read_initial_energy(ep0_json)

    # --- figure size to match example (130pt  110pt) ---
    fig, ax = plt.subplots(figsize=(pt_to_in(130), pt_to_in(110)), dpi=150)
    loaded_any = False
    max_step = 0

    # ---- SA curve (per-step) ----
    if sa_csv and sa_csv.exists():
        try:
            xs, ys = read_sa_step_curve(sa_csv, initial_energy)
            if len(xs) > 0:
                max_step = int(np.max(xs))
                ax.plot(xs, ys, label=SA_LABEL, color=SA_COLOR, linestyle='-', linewidth=2)
                loaded_any = True
        except Exception as e:
            print(f"[{hex_id}] Failed to read SA CSV {sa_csv}: {e}")
    else:
        print(f"[{hex_id}] Missing SA metrics CSV: {sa_csv}")

    # ---- optional dashed ML baseline ----
    ml_size = size_lookup.get(_norm_hex(hex_id))
    if pd.notna(ml_size):
        ax.axhline(
            y=float(ml_size),
            linestyle="--",
            linewidth=0.5,
            color="#444444",
            label="DeepCirc trained from scratch on single design",
        )

    if loaded_any:
        # Aesthetics like the example
        for spine in ax.spines.values():
            spine.set_linewidth(0.5)

        right_limit = int(np.ceil(max_step / BASE_TICK_STEPS) * BASE_TICK_STEPS) if max_step > 0 else BASE_TICK_STEPS
        ax.xaxis.set_major_locator(mticker.MultipleLocator(BASE_TICK_STEPS))
        # ax.set_xlim(0, right_limit)
        ax.set_xlim(-0.02 * max(right_limit, 1), right_limit)  # tiny left pad like before
        ax.margins(x=0.02)
        ax.minorticks_off()
        ax.tick_params(axis='both', which='both', direction='out', length=2, width=0.5)

        ax.set_xlabel("Environment steps")
        ax.set_ylabel("Number of repressors")
        title_suffix = " (step0 = initial from JSON)" if initial_energy is not None else ""
        ax.set_title(f"Circuit {hex_id}{title_suffix}")

        ax.legend(loc="center left", bbox_to_anchor=(1.02, 0.5), borderaxespad=0)
        plt.tight_layout(rect=[0, 0, 0.85, 1])
        plt.show()
    else:
        print(f"[{hex_id}] No curves could be plotted (missing/invalid CSVs).")
```

SA (11,400) vs trained from scratch (new) (>50,000)

```python
# ---------- SA vs SCRATCH (TensorBoard) ----------
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from typing import Dict, List, Tuple, Optional
from tensorboard.backend.event_processing import event_accumulator

# ====== you already have these in scope ======
# hex_ids: List[str]

# ---- TB config for SCRATCH ----
SCRATCH_ROOT_TEMPLATE = (
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/scratch_training/{HEX}/seed_1"
)
EVENT_GLOBS = ["**/events.out.tfevents.*", "**/*tfevents*"]
TB_TAG = "custom/best_energy"
ASSUME_TAG_IS_BEST = True  # True: compare final point; False: compare global min over series

# ---- Normalization ----
def _norm_hex(x: str) -> str:
    s = str(x).strip().lower()
    if not s.startswith("0x"):
        s = "0x" + s
    return "0x" + s[2:].upper()

# ---- SA CSV discovery patterns ----
METHODS = ["sa"]
LABELS = {"sa": "Simulated Annealing (masked + backtrack)"}
patterns = {
    "sa": [
        "**/sa_masked/sa_step_metrics.csv",
        "**/sa_masked/*sa_step_metrics.csv",
        "**/*sa_step_metrics.csv",
    ],
}

def find_csv(search_root: Path, method: str) -> Optional[Path]:
    root = Path(search_root)
    for pat in patterns[method]:
        matches = sorted(set(p.as_posix() for p in root.glob(pat)))
        if matches:
            return Path(matches[0])
    # fallback if saved at root
    if method == "sa":
        ex = root / "sa_step_metrics.csv"
        if ex.exists():
            return ex
    return None

def get_final_size(csv_path: Path) -> Optional[float]:
    """
    For SA per-step metrics:
      - expects columns: step, energy, [best_energy_so_far]
      - returns the final best_energy_so_far (or cummin(energy) at last step)
    """
    try:
        df = pd.read_csv(csv_path)
        # Ensure numeric and sorted by step if present
        if "step" in df.columns:
            df["step"] = pd.to_numeric(df["step"], errors="coerce")
            df = df.dropna(subset=["step"]).sort_values("step", kind="mergesort")
        if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
            s = pd.to_numeric(df["best_energy_so_far"], errors="coerce").dropna()
            if len(s) > 0:
                return float(s.iloc[-1])
        # Fallback: compute cumulative min of energy
        if "energy" in df.columns and df["energy"].notna().any():
            e = pd.to_numeric(df["energy"], errors="coerce").dropna()
            if len(e) > 0:
                return float(e.cummin().iloc[-1])
    except Exception as e:
        print(f"[WARN] Failed to read {csv_path}: {e}")
    return None

# ---- SCRATCH TensorBoard helpers ----
def scratch_root_for_hex(hex_id: str) -> Path:
    return Path(SCRATCH_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

def find_event_files(root: Path) -> List[Path]:
    files = []
    for pat in EVENT_GLOBS:
        files.extend(root.glob(pat))
    return sorted(set(files), key=lambda p: p.as_posix())

def load_tb_scalar_from_dir(run_dir: Path, tag: str) -> Tuple[np.ndarray, np.ndarray]:
    try:
        ea = event_accumulator.EventAccumulator(str(run_dir))
        ea.Reload()
        if tag not in ea.Tags().get("scalars", []):
            return np.array([], dtype=int), np.array([], dtype=float)
        events = ea.Scalars(tag)
        steps  = np.fromiter((e.step  for e in events), dtype=int)
        values = np.fromiter((e.value for e in events), dtype=float)
        return steps, values
    except Exception as e:
        print(f"[WARN] TB load failed in {run_dir}: {e}")
        return np.array([], dtype=int), np.array([], dtype=float)

def pick_best_from_runs(run_dirs: List[Path], tag: str, assume_best_series=True) -> Tuple[Optional[float], Optional[Path], np.ndarray, np.ndarray]:
    """Return (best_value, chosen_dir, steps, values) for the best run (lowest value)."""
    best_val, best_dir, best_steps, best_values = None, None, None, None
    for d in run_dirs:
        steps, vals = load_tb_scalar_from_dir(d, tag)
        if steps.size == 0:
            continue
        candidate = float(vals[-1]) if assume_best_series else float(np.nanmin(vals))
        if best_val is None or candidate < best_val:
            best_val, best_dir, best_steps, best_values = candidate, d, steps, vals
    return best_val, best_dir, best_steps, best_values

# ======== PLOTTING: creation only (no save/show here) ========
def make_bar_figure(categories: List[str], values: List[float], title: str) -> plt.Figure:
    fig, ax = plt.subplots(figsize=(4, 3))
    bars = ax.bar(categories, values)
    ax.set_ylim(0, 100)
    ax.set_ylabel("Percentage of circuits")
    ax.set_title(title)
    for rect, v in zip(bars, values):
        ax.text(
            rect.get_x() + rect.get_width() / 2,
            rect.get_height() + 1,
            f"{v:.1f}%",
            ha="center",
            va="bottom",
            fontsize=9,
        )
    fig.tight_layout()
    return fig

def save_fig(fig: plt.Figure, path: Path, dpi: int = 200):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    fig.savefig(path, dpi=dpi, bbox_inches="tight")

def show_fig(fig: plt.Figure):
    plt.show()

def close_fig(fig: plt.Figure):
    plt.close(fig)

def to_percentages(counts: Dict[str, int], denom: int) -> List[float]:
    if denom == 0:
        return [0.0, 0.0, 0.0]
    return [
        100.0 * counts["Win"] / denom,
        100.0 * counts["Loss"] / denom,
        100.0 * counts["Tie"] / denom,
    ]

# ======== Core comparison: SA vs SCRATCH ========
def _sa_root_for_hex(hex_id: str) -> Path:
    # adjust if your SA results live elsewhere
    return Path(
        f"/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
        f"manuscript/logic_synthesis_SA/11400_steps/4_inputs/no_fw_preference/{_norm_hex(hex_id)}/seed_1"
    )

def compare_sa_to_scratch(
    hex_ids: List[str],
    tb_tag: str = TB_TAG,
    assume_tag_is_best: bool = ASSUME_TAG_IS_BEST,
    return_rows: bool = True,
):
    outcomes = {"Win": 0, "Loss": 0, "Tie": 0}
    n = 0
    missing = []
    rows = []

    for hex_id in hex_ids:
        hex_norm = _norm_hex(hex_id)

        # SA (CSV)
        sa_root = _sa_root_for_hex(hex_norm)
        sa_csv = find_csv(sa_root, "sa")
        sa_final = get_final_size(sa_csv) if sa_csv else None

        # SCRATCH (TensorBoard)
        sc_root = scratch_root_for_hex(hex_norm)
        sc_events = find_event_files(sc_root)
        sc_run_dirs = sorted(set(p.parent for p in sc_events))
        sc_best, sc_dir, sc_steps, sc_vals = pick_best_from_runs(sc_run_dirs, tb_tag, assume_tag_is_best)

        if sa_final is None or sc_best is None:
            missing.append(hex_norm)
            continue

        n += 1
        if sa_final < sc_best:
            outcomes["Win"] += 1; verdict = "Win"
        elif sa_final > sc_best:
            outcomes["Loss"] += 1; verdict = "Loss"
        else:
            outcomes["Tie"] += 1; verdict = "Tie"

        if return_rows:
            rows.append({
                "hex_id": hex_norm,
                "sa_final": float(sa_final),
                "scratch_best": float(sc_best),
                "delta_sa_minus_scratch": float(sa_final - sc_best),
                "result": verdict,
                "sa_csv": str(sa_csv) if sa_csv else "",
                "scratch_chosen_run_dir": str(sc_dir) if sc_dir else "",
            })

    return outcomes, n, missing, (pd.DataFrame(rows) if return_rows else None)

# ======== Orchestrator (SA vs SCRATCH) ========
def summarize_and_plot_sa_vs_scratch(
    hex_ids: List[str],
    tb_tag: str = TB_TAG,
    assume_tag_is_best: bool = ASSUME_TAG_IS_BEST,
    make_plots: bool = True,
    show_plots: bool = False,
    save_summary_csv: Optional[Path] = None,
    save_plot_path: Optional[Path] = None,
):
    """
    Returns (summary_df, per_hex_df, fig_or_None)
    """
    cats = ["Win", "Loss", "Tie"]
    outcomes, n_used, missing, per_hex_df = compare_sa_to_scratch(
        hex_ids, tb_tag, assume_tag_is_best, return_rows=True
    )
    pct = to_percentages(outcomes, n_used)
    print(f"SA vs SCRATCH  counts: {outcomes} out of N={n_used} (missing {len(missing)})")

    summary = pd.DataFrame([{
        "method": "sa_vs_scratch",
        "label": "Simulated Annealing vs Trained-from-Scratch",
        "N_used": n_used,
        "Win": outcomes["Win"],
        "Loss": outcomes["Loss"],
        "Tie": outcomes["Tie"],
        "Win_%": pct[0],
        "Loss_%": pct[1],
        "Tie_%": pct[2],
        "N_missing": len(missing),
        "Missing_hex_ids": ",".join(sorted(missing)),
        "tb_tag": tb_tag,
        "assume_tag_is_best": assume_tag_is_best,
    }])

    fig = None
    if make_plots:
        title = f"SA vs Scratch (N={n_used})  tag: {tb_tag}"
        fig = make_bar_figure(cats, pct, title)
        if save_plot_path is not None:
            save_fig(fig, Path(save_plot_path))
        if show_plots:
            show_fig(fig)

    if save_summary_csv is not None:
        Path(save_summary_csv).parent.mkdir(parents=True, exist_ok=True)
        summary.to_csv(save_summary_csv, index=False)
        # also save per-hex details
        per_hex_out = save_summary_csv.with_name(save_summary_csv.stem + "_per_hex.csv")
        if per_hex_df is not None:
            per_hex_df.to_csv(per_hex_out, index=False)

    return summary, per_hex_df, fig

# ======================== example usage ========================
summary_df, per_hex_df, fig = summarize_and_plot_sa_vs_scratch(
    hex_ids,
    tb_tag="custom/best_energy",   # or your desired TB scalar tag
    assume_tag_is_best=True,       # True: compare last point; False: compare global min
    make_plots=True,
    show_plots=True,
    # save_summary_csv=Path("./sa_vs_scratch_summary.csv"),
    # save_plot_path=Path("./sa_vs_scratch.png"),
)
```

```python
# ---------- SA vs SCRATCH (TensorBoard) ----------
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from typing import Dict, List, Tuple, Optional
from tensorboard.backend.event_processing import event_accumulator

# ====== you already have these in scope ======
# hex_ids: List[str]

# ---- TB config for SCRATCH ----
SCRATCH_ROOT_TEMPLATE = (
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/fine_tune/GAT_MLP_with_scalars/4000nn/initial_state_sampling_factor_0/4_inputs/50_000_steps/{HEX}/seed_1"
)
EVENT_GLOBS = ["**/events.out.tfevents.*", "**/*tfevents*"]
TB_TAG = "custom/best_energy"
ASSUME_TAG_IS_BEST = True  # True: compare final point; False: compare global min over series

# ---- Normalization ----
def _norm_hex(x: str) -> str:
    s = str(x).strip().lower()
    if not s.startswith("0x"):
        s = "0x" + s
    return "0x" + s[2:].upper()

# ---- SA CSV discovery patterns ----
METHODS = ["sa"]
LABELS = {"sa": "Simulated Annealing (masked + backtrack)"}
patterns = {
    "sa": [
        "**/sa_masked/sa_step_metrics.csv",
        "**/sa_masked/*sa_step_metrics.csv",
        "**/*sa_step_metrics.csv",
    ],
}

def find_csv(search_root: Path, method: str) -> Optional[Path]:
    root = Path(search_root)
    for pat in patterns[method]:
        matches = sorted(set(p.as_posix() for p in root.glob(pat)))
        if matches:
            return Path(matches[0])
    # fallback if saved at root
    if method == "sa":
        ex = root / "sa_step_metrics.csv"
        if ex.exists():
            return ex
    return None

def get_final_size(csv_path: Path) -> Optional[float]:
    """
    For SA per-step metrics:
      - expects columns: step, energy, [best_energy_so_far]
      - returns the final best_energy_so_far (or cummin(energy) at last step)
    """
    try:
        df = pd.read_csv(csv_path)
        # Ensure numeric and sorted by step if present
        if "step" in df.columns:
            df["step"] = pd.to_numeric(df["step"], errors="coerce")
            df = df.dropna(subset=["step"]).sort_values("step", kind="mergesort")
        if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
            s = pd.to_numeric(df["best_energy_so_far"], errors="coerce").dropna()
            if len(s) > 0:
                return float(s.iloc[-1])
        # Fallback: compute cumulative min of energy
        if "energy" in df.columns and df["energy"].notna().any():
            e = pd.to_numeric(df["energy"], errors="coerce").dropna()
            if len(e) > 0:
                return float(e.cummin().iloc[-1])
    except Exception as e:
        print(f"[WARN] Failed to read {csv_path}: {e}")
    return None

# ---- SCRATCH TensorBoard helpers ----
def scratch_root_for_hex(hex_id: str) -> Path:
    return Path(SCRATCH_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

def find_event_files(root: Path) -> List[Path]:
    files = []
    for pat in EVENT_GLOBS:
        files.extend(root.glob(pat))
    return sorted(set(files), key=lambda p: p.as_posix())

def load_tb_scalar_from_dir(run_dir: Path, tag: str) -> Tuple[np.ndarray, np.ndarray]:
    try:
        ea = event_accumulator.EventAccumulator(str(run_dir))
        ea.Reload()
        if tag not in ea.Tags().get("scalars", []):
            return np.array([], dtype=int), np.array([], dtype=float)
        events = ea.Scalars(tag)
        steps  = np.fromiter((e.step  for e in events), dtype=int)
        values = np.fromiter((e.value for e in events), dtype=float)
        return steps, values
    except Exception as e:
        print(f"[WARN] TB load failed in {run_dir}: {e}")
        return np.array([], dtype=int), np.array([], dtype=float)

def pick_best_from_runs(run_dirs: List[Path], tag: str, assume_best_series=True) -> Tuple[Optional[float], Optional[Path], np.ndarray, np.ndarray]:
    """Return (best_value, chosen_dir, steps, values) for the best run (lowest value)."""
    best_val, best_dir, best_steps, best_values = None, None, None, None
    for d in run_dirs:
        steps, vals = load_tb_scalar_from_dir(d, tag)
        if steps.size == 0:
            continue
        candidate = float(vals[-1]) if assume_best_series else float(np.nanmin(vals))
        if best_val is None or candidate < best_val:
            best_val, best_dir, best_steps, best_values = candidate, d, steps, vals
    return best_val, best_dir, best_steps, best_values

# ======== PLOTTING: creation only (no save/show here) ========
def make_bar_figure(categories: List[str], values: List[float], title: str) -> plt.Figure:
    fig, ax = plt.subplots(figsize=(4, 3))
    bars = ax.bar(categories, values)
    ax.set_ylim(0, 100)
    ax.set_ylabel("Percentage of circuits")
    ax.set_title(title)
    for rect, v in zip(bars, values):
        ax.text(
            rect.get_x() + rect.get_width() / 2,
            rect.get_height() + 1,
            f"{v:.1f}%",
            ha="center",
            va="bottom",
            fontsize=9,
        )
    fig.tight_layout()
    return fig

def save_fig(fig: plt.Figure, path: Path, dpi: int = 200):
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)
    fig.savefig(path, dpi=dpi, bbox_inches="tight")

def show_fig(fig: plt.Figure):
    plt.show()

def close_fig(fig: plt.Figure):
    plt.close(fig)

def to_percentages(counts: Dict[str, int], denom: int) -> List[float]:
    if denom == 0:
        return [0.0, 0.0, 0.0]
    return [
        100.0 * counts["Win"] / denom,
        100.0 * counts["Loss"] / denom,
        100.0 * counts["Tie"] / denom,
    ]

# ======== Core comparison: SA vs SCRATCH ========
def _sa_root_for_hex(hex_id: str) -> Path:
    # adjust if your SA results live elsewhere
    return Path(
        f"/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
        f"manuscript/logic_synthesis_SA/11400_steps/4_inputs/fw_preference/{_norm_hex(hex_id)}/seed_1"
    )

def compare_sa_to_scratch(
    hex_ids: List[str],
    tb_tag: str = TB_TAG,
    assume_tag_is_best: bool = ASSUME_TAG_IS_BEST,
    return_rows: bool = True,
):
    outcomes = {"Win": 0, "Loss": 0, "Tie": 0}
    n = 0
    missing = []
    rows = []

    for hex_id in hex_ids:
        hex_norm = _norm_hex(hex_id)

        # SA (CSV)
        sa_root = _sa_root_for_hex(hex_norm)
        sa_csv = find_csv(sa_root, "sa")
        sa_final = get_final_size(sa_csv) if sa_csv else None

        # SCRATCH (TensorBoard)
        sc_root = scratch_root_for_hex(hex_norm)
        sc_events = find_event_files(sc_root)
        sc_run_dirs = sorted(set(p.parent for p in sc_events))
        sc_best, sc_dir, sc_steps, sc_vals = pick_best_from_runs(sc_run_dirs, tb_tag, assume_tag_is_best)

        if sa_final is None or sc_best is None:
            missing.append(hex_norm)
            continue

        n += 1
        if sa_final < sc_best:
            outcomes["Win"] += 1; verdict = "Win"
        elif sa_final > sc_best:
            outcomes["Loss"] += 1; verdict = "Loss"
        else:
            outcomes["Tie"] += 1; verdict = "Tie"

        if return_rows:
            rows.append({
                "hex_id": hex_norm,
                "sa_final": float(sa_final),
                "scratch_best": float(sc_best),
                "delta_sa_minus_scratch": float(sa_final - sc_best),
                "result": verdict,
                "sa_csv": str(sa_csv) if sa_csv else "",
                "scratch_chosen_run_dir": str(sc_dir) if sc_dir else "",
            })

    return outcomes, n, missing, (pd.DataFrame(rows) if return_rows else None)

# ======== Orchestrator (SA vs SCRATCH) ========
def summarize_and_plot_sa_vs_scratch(
    hex_ids: List[str],
    tb_tag: str = TB_TAG,
    assume_tag_is_best: bool = ASSUME_TAG_IS_BEST,
    make_plots: bool = True,
    show_plots: bool = False,
    save_summary_csv: Optional[Path] = None,
    save_plot_path: Optional[Path] = None,
):
    """
    Returns (summary_df, per_hex_df, fig_or_None)
    """
    cats = ["Win", "Loss", "Tie"]
    outcomes, n_used, missing, per_hex_df = compare_sa_to_scratch(
        hex_ids, tb_tag, assume_tag_is_best, return_rows=True
    )
    pct = to_percentages(outcomes, n_used)
    print(f"SA vs SCRATCH  counts: {outcomes} out of N={n_used} (missing {len(missing)})")

    summary = pd.DataFrame([{
        "method": "sa_vs_scratch",
        "label": "Simulated Annealing vs Trained-from-Scratch",
        "N_used": n_used,
        "Win": outcomes["Win"],
        "Loss": outcomes["Loss"],
        "Tie": outcomes["Tie"],
        "Win_%": pct[0],
        "Loss_%": pct[1],
        "Tie_%": pct[2],
        "N_missing": len(missing),
        "Missing_hex_ids": ",".join(sorted(missing)),
        "tb_tag": tb_tag,
        "assume_tag_is_best": assume_tag_is_best,
    }])

    fig = None
    if make_plots:
        title = f"SA vs Scratch (N={n_used})  tag: {tb_tag}"
        fig = make_bar_figure(cats, pct, title)
        if save_plot_path is not None:
            save_fig(fig, Path(save_plot_path))
        if show_plots:
            show_fig(fig)

    if save_summary_csv is not None:
        Path(save_summary_csv).parent.mkdir(parents=True, exist_ok=True)
        summary.to_csv(save_summary_csv, index=False)
        # also save per-hex details
        per_hex_out = save_summary_csv.with_name(save_summary_csv.stem + "_per_hex.csv")
        if per_hex_df is not None:
            per_hex_df.to_csv(per_hex_out, index=False)

    return summary, per_hex_df, fig

# ======================== example usage ========================
summary_df, per_hex_df, fig = summarize_and_plot_sa_vs_scratch(
    hex_ids,
    tb_tag="custom/best_energy",   # or your desired TB scalar tag
    assume_tag_is_best=True,       # True: compare last point; False: compare global min
    make_plots=True,
    show_plots=True,
    # save_summary_csv=Path("./sa_vs_scratch_summary.csv"),
    # save_plot_path=Path("./sa_vs_scratch.png"),
)
```

```python
# Fine-tuning (TensorBoard) vs SCRATCH (TensorBoard)
# - Prints per-circuit Wins/Losses/Ties by hex_id

from tensorboard.backend.event_processing import event_accumulator
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.ticker as mticker
from typing import Dict, List, Tuple, Optional

# ----------------
mpl.rcParams.update({
    "font.family": "DejaVu Sans",
    "font.size": 5, "axes.titlesize": 5, "axes.labelsize": 5,
    "legend.fontsize": 5, "xtick.labelsize": 5, "ytick.labelsize": 5,
    "pdf.fonttype": 42, "ps.fonttype": 42,
    "pdf.use14corefonts": False, "svg.fonttype": "none",
    "axes.linewidth": 0.5, "lines.linewidth": 0.5,
    "legend.frameon": False,
    "savefig.dpi": 300, "figure.dpi": 100,
})
EDGE_COL, EDGE_W = "#58595B", 0.5
pt_to_in = lambda pt: pt / 72.0

def _style_axes(ax, grid_y=True):
    # ticks
    ax.tick_params(axis="y", which="major", width=EDGE_W, length=6, color=EDGE_COL)
    ax.tick_params(axis="y", which="minor", width=EDGE_W, length=3, color=EDGE_COL)
    ax.tick_params(axis="x", which="both",  width=EDGE_W, length=3, color=EDGE_COL)
    # spines
    for s in ["left", "bottom"]:
        ax.spines[s].set_linewidth(EDGE_W)
        ax.spines[s].set_color(EDGE_COL)
    for s in ["top", "right"]:
        ax.spines[s].set_visible(False)
    # grid
    if grid_y:
        ax.grid(True, axis="y", which="major", linewidth=0.3, alpha=0.3)

SCRATCH_ROOT_TEMPLATE = (
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/scratch_training/{HEX}/seed_1"
)

FINETUNE_ROOT_TEMPLATE = (
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/fine_tune/GAT_MLP_with_scalars/4000nn/initial_state_sampling_factor_0/"
    "4_inputs/50_000_steps/{HEX}/seed_1"
)

EVENT_GLOBS = ["**/events.out.tfevents.*", "**/*tfevents*"]
TB_TAG = "custom/best_energy"
ASSUME_TAG_IS_BEST = True  # if True, use final point; else use global min over series

def _norm_hex(x: str) -> str:
    s = str(x).strip().lower()
    if not s.startswith("0x"):
        s = "0x" + s
    return "0x" + s[2:].upper()

def scratch_root_for_hex(hex_id: str) -> Path:
    return Path(SCRATCH_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

def finetune_root_for_hex(hex_id: str) -> Path:
    return Path(FINETUNE_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

def find_event_files(root: Path) -> List[Path]:
    files = []
    for pat in EVENT_GLOBS:
        files.extend(root.glob(pat))
    return sorted(set(files), key=lambda p: p.as_posix())

def load_tb_scalar_from_dir(run_dir: Path, tag: str) -> Tuple[np.ndarray, np.ndarray]:
    try:
        ea = event_accumulator.EventAccumulator(str(run_dir))
        ea.Reload()
        if tag not in ea.Tags().get("scalars", []):
            return np.array([], dtype=int), np.array([], dtype=float)
        events = ea.Scalars(tag)
        steps  = np.fromiter((e.step  for e in events), dtype=int)
        values = np.fromiter((e.value for e in events), dtype=float)
        return steps, values
    except Exception as e:
        print(f"[WARN] TB load failed in {run_dir}: {e}")
        return np.array([], dtype=int), np.array([], dtype=float)

def pick_best_from_runs(run_dirs: List[Path], tag: str, assume_best_series=True) -> Tuple[Optional[float], Optional[Path], np.ndarray, np.ndarray]:
    """Return (best_value, chosen_dir, steps, values) for the best run (lowest value)."""
    best_val, best_dir, best_steps, best_values = None, None, None, None
    for d in run_dirs:
        steps, vals = load_tb_scalar_from_dir(d, tag)
        if steps.size == 0:
            continue
        final = float(vals[-1]) if assume_best_series else float(np.nanmin(vals))
        if best_val is None or final < best_val:
            best_val, best_dir, best_steps, best_values = final, d, steps, vals
    return best_val, best_dir, best_steps, best_values

def _steps_to_global_min(steps: Optional[np.ndarray], vals: Optional[np.ndarray]) -> float:
    """Return the step index where the (first) global minimum is reached, or NaN."""
    if steps is None or vals is None or len(steps) == 0 or len(vals) == 0:
        return float("nan")
    idx = int(np.nanargmin(vals))
    return float(steps[idx])


def compare_finetune_to_scratch_tb(
    hex_ids: List[str],
    tag: str = TB_TAG,
    assume_tag_is_best: bool = ASSUME_TAG_IS_BEST,
    make_plot: bool = True,
    show_plot: bool = True,
    save_plot_path: Optional[Path] = None,
    save_summary_csv: Optional[Path] = None,
    per_hex_plot_dir: Optional[Path] = None,
    # Existing NEW:
    make_win_comparison_plot: bool = True,
    save_win_comparison_plot_path: Optional[Path] = None,
    # NEW (this request):
    make_tie_steps_plot: bool = True,
    save_tie_steps_plot_path: Optional[Path] = None,
):
    outcomes = {"Win": 0, "Loss": 0, "Tie": 0}
    rows = []
    missing = []

    # Track exactly which circuits fall into each bucket
    win_hex, loss_hex, tie_hex = [], [], []

    for hex_id in hex_ids:
        hex_norm = _norm_hex(hex_id)

        # FINETUNE
        ft_root = finetune_root_for_hex(hex_norm)
        ft_events = find_event_files(ft_root)
        ft_run_dirs = sorted(set(p.parent for p in ft_events))
        ft_best, ft_dir, ft_steps, ft_vals = pick_best_from_runs(ft_run_dirs, tag, assume_tag_is_best)

        # SCRATCH
        sc_root = scratch_root_for_hex(hex_norm)
        sc_events = find_event_files(sc_root)
        sc_run_dirs = sorted(set(p.parent for p in sc_events))
        sc_best, sc_dir, sc_steps, sc_vals = pick_best_from_runs(sc_run_dirs, tag, assume_tag_is_best)

        if ft_best is None or sc_best is None:
            missing.append(hex_norm)
            continue

        if ft_best < sc_best:
            outcomes["Win"] += 1; verdict = "Win";  win_hex.append(hex_norm)
        elif ft_best > sc_best:
            outcomes["Loss"] += 1; verdict = "Loss"; loss_hex.append(hex_norm)
        else:
            outcomes["Tie"] += 1; verdict = "Tie";  tie_hex.append(hex_norm)

        # Steps to *global min* (when the best energy was first achieved)
        ft_steps_to_best = _steps_to_global_min(ft_steps, ft_vals)
        sc_steps_to_best = _steps_to_global_min(sc_steps, sc_vals)

        rows.append({
            "hex_id": hex_norm,
            "finetune_best": float(ft_best),
            "scratch_best": float(sc_best),
            "delta_finetune_minus_scratch": float(ft_best - sc_best),
            "result": verdict,
            "finetune_chosen_run_dir": str(ft_dir) if ft_dir else "",
            "scratch_chosen_run_dir": str(sc_dir) if sc_dir else "",
            # NEW columns for downstream analysis and the tie-steps plot
            "finetune_steps_to_best": ft_steps_to_best,
            "scratch_steps_to_best": sc_steps_to_best,
        })

        # Optional per-hex plot of both series (kept simple; export style mainly for summary figs)
        if per_hex_plot_dir is not None and ft_steps is not None and ft_vals is not None:
            per_hex_plot_dir = Path(per_hex_plot_dir)
            per_hex_plot_dir.mkdir(parents=True, exist_ok=True)
            fig_hex, ax = plt.subplots(figsize=(pt_to_in(230), pt_to_in(170)), dpi=300)
            if sc_steps is not None and sc_vals is not None and len(sc_steps) > 0:
                ax.plot(sc_steps, sc_vals, lw=1.0, label="scratch")
                ax.axhline(y=float(sc_best), color=EDGE_COL, ls="--", lw=EDGE_W, label="scratch best")
            ax.plot(ft_steps, ft_vals, lw=1.0, label="finetune")
            ax.axhline(y=float(ft_best), color=EDGE_COL, ls=":", lw=EDGE_W, label="finetune best")
            ax.set_xlabel("Training steps")
            ax.set_ylabel("Metric (lower is better)")
            ax.set_title(f"{hex_norm}  (FT={ft_best:.3g}, SC={sc_best:.3g})")
            ax.legend(frameon=False)
            _style_axes(ax, grid_y=True)
            fig_hex.tight_layout()
            fig_hex.savefig(per_hex_plot_dir / f"{hex_norm}.png", bbox_inches="tight")
            plt.close(fig_hex)

    n_used = len(rows)
    print(f"[Fine-tune vs Scratch] counts: {outcomes} out of N={n_used} (missing {len(missing)})")
    if win_hex:
        print(f"  Wins  ({len(win_hex)}): {', '.join(sorted(win_hex))}")
    if loss_hex:
        print(f"  Losses({len(loss_hex)}): {', '.join(sorted(loss_hex))}")
    if tie_hex:
        print(f"  Ties  ({len(tie_hex)}): {', '.join(sorted(tie_hex))}")
    if missing:
        print(f"  Missing ({len(missing)}): {', '.join(sorted(missing))}")

    per_hex = pd.DataFrame(rows)
    if n_used == 0:
        win_pct = loss_pct = tie_pct = 0.0
    else:
        win_pct  = 100.0 * outcomes["Win"]  / n_used
        loss_pct = 100.0 * outcomes["Loss"] / n_used
        tie_pct  = 100.0 * outcomes["Tie"]  / n_used

    summary = pd.DataFrame([{
        "N_used": n_used,
        "Win": outcomes["Win"], "Loss": outcomes["Loss"], "Tie": outcomes["Tie"],
        "Win_%": win_pct, "Loss_%": loss_pct, "Tie_%": tie_pct,
        "N_missing": len(missing),
        "Missing_hex_ids": ",".join(sorted(missing)),
        "Wins_hex_ids": ",".join(sorted(win_hex)),
        "Losses_hex_ids": ",".join(sorted(loss_hex)),
        "Ties_hex_ids": ",".join(sorted(tie_hex)),
        "tag": tag,
        "assume_tag_is_best": assume_tag_is_best,
    }])

    # --- Figure 1: overall Win/Loss/Tie percentages (styled) ---
    fig = None
    if make_plot:
        cats = ["Win", "Loss", "Tie"]
        vals = [win_pct, loss_pct, tie_pct]
        fig, ax = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
        bars = ax.bar(cats, vals,
                      color=["#4CAF50", "#F44336", "#9E9E9E"],
                      edgecolor=EDGE_COL, linewidth=EDGE_W)
        ax.set_ylim(0, 100)
        ax.set_ylabel("Percentage of circuits")
        ax.set_title(f"Fine-tune vs Scratch  Quality (N={n_used})")
        for i, v in enumerate(vals):
            ax.text(i, min(100, v) + 0.5, f"{v:.1f}%", ha="center", va="bottom")
        _style_axes(ax, grid_y=True)
        fig.tight_layout()
        if save_plot_path is not None:
            save_plot_path = Path(save_plot_path)
            save_plot_path.parent.mkdir(parents=True, exist_ok=True)
            fig.savefig(save_plot_path, bbox_inches="tight")
        if show_plot:
            plt.show()

    # --- Figure 2: comparison among wins only (styled) ---
    fig_win = None
    if make_win_comparison_plot:
        wins_only = per_hex[per_hex["result"] == "Win"].copy()
        if len(wins_only) > 0:
            fig_win, ax2 = plt.subplots(figsize=(pt_to_in(100), pt_to_in(100)), dpi=300)
            x_sc = wins_only["scratch_best"].to_numpy(dtype=float)
            y_ft = wins_only["finetune_best"].to_numpy(dtype=float)
            ax2.scatter(x_sc, y_ft, s=8, edgecolors=EDGE_COL, linewidths=EDGE_W/2)
            lo = float(min(np.nanmin(x_sc), np.nanmin(y_ft)))
            hi = float(max(np.nanmax(x_sc), np.nanmax(y_ft)))
            pad = 0.02 * (hi - lo) if hi > lo else 1.0
            ax2.plot([lo - pad, hi + pad], [lo - pad, hi + pad], "--", lw=EDGE_W, color=EDGE_COL)
            ax2.set_xlabel("Scratch best (lower is better)")
            ax2.set_ylabel("Fine-tune best (lower is better)")
            ax2.set_title(f"Where Fine-tuning Beat Scratch (N={len(wins_only)})")
            ax2.set_aspect("equal", adjustable="box")
            _style_axes(ax2, grid_y=True)
            fig_win.tight_layout()
            if save_win_comparison_plot_path is not None:
                save_win_comparison_plot_path = Path(save_win_comparison_plot_path)
                save_win_comparison_plot_path.parent.mkdir(parents=True, exist_ok=True)
                fig_win.savefig(save_win_comparison_plot_path, bbox_inches="tight")
            if show_plot:
                plt.show()
        else:
            print("[Info] No winning circuits; skipping win-comparison plot.")

    # --- Figure 3 (NEW): Ties  steps to reach best energy (styled) ---
    fig_tie_steps = None
    if make_tie_steps_plot:
        ties_only = per_hex[per_hex["result"] == "Tie"].copy()
        ties_only = ties_only.dropna(subset=["scratch_steps_to_best", "finetune_steps_to_best"])
        if len(ties_only) > 0:
            fig_tie_steps, ax3 = plt.subplots(figsize=(pt_to_in(100), pt_to_in(100)), dpi=300)
            x_sc = ties_only["scratch_steps_to_best"].astype(float).to_numpy()
            y_ft = ties_only["finetune_steps_to_best"].astype(float).to_numpy()
            ax3.scatter(x_sc, y_ft, s=8, edgecolors=EDGE_COL, linewidths=EDGE_W/2)
            lo = float(min(np.nanmin(x_sc), np.nanmin(y_ft)))
            hi = float(max(np.nanmax(x_sc), np.nanmax(y_ft)))
            pad = 0.02 * (hi - lo) if hi > lo else 1.0
            ax3.plot([lo - pad, hi + pad], [lo - pad, hi + pad], "--", lw=EDGE_W, color=EDGE_COL)
            ax3.set_xlabel("Scratch steps to best (lower is faster)")
            ax3.set_ylabel("Fine-tune steps to best (lower is faster)")
            ax3.set_title(f"Ties: Steps to Reach Best Energy (N={len(ties_only)})")
            ax3.set_aspect("equal", adjustable="box")
            _style_axes(ax3, grid_y=True)
            fig_tie_steps.tight_layout()
            if save_tie_steps_plot_path is not None:
                save_tie_steps_plot_path = Path(save_tie_steps_plot_path)
                save_tie_steps_plot_path.parent.mkdir(parents=True, exist_ok=True)
                fig_tie_steps.savefig(save_tie_steps_plot_path, bbox_inches="tight")
            if show_plot:
                plt.show()
        else:
            print("[Info] No ties; skipping tie-steps plot.")

    if save_summary_csv is not None:
        save_summary_csv = Path(save_summary_csv)
        save_summary_csv.parent.mkdir(parents=True, exist_ok=True)
        summary.to_csv(save_summary_csv, index=False)
        per_hex_path = save_summary_csv.with_name(save_summary_csv.stem + "_per_hex.csv")
        per_hex.to_csv(per_hex_path, index=False)

    # Return all figures
    return summary, per_hex, fig, fig_win, fig_tie_steps



# ---- Example call ----
summary_df, per_hex_df, fig, fig_win, fig_tie_steps = compare_finetune_to_scratch_tb(
    hex_ids=hex_ids,                  # list of "0x.." ids
    tag="custom/best_energy",
    assume_tag_is_best=True,          # True: use final; False: global min
    make_plot=True,
    show_plot=True,
    make_win_comparison_plot=True,
    # save_plot_path=Path("./finetune_vs_scratch.png"),
    # save_win_comparison_plot_path=Path("./finetune_wins_scatter.png"),
    # save_summary_csv=Path("./finetune_vs_scratch_summary.csv"),
    # per_hex_plot_dir=Path("./per_hex_plots"),
)

```

```python
from tensorboard.backend.event_processing import event_accumulator
import re
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.patches import Patch
from typing import Dict, List, Tuple, Optional

# ---------------- Unified, export-friendly style (5 pt text) ----------------
mpl.rcParams.update({
    "font.size": 5, "axes.titlesize": 5, "axes.labelsize": 5,
    "legend.fontsize": 5, "xtick.labelsize": 5, "ytick.labelsize": 5,
    "pdf.fonttype": 42, "ps.fonttype": 42,
    "pdf.use14corefonts": False, "svg.fonttype": "none",
    "axes.linewidth": 0.5, "lines.linewidth": 0.5,
    "legend.frameon": False,
    "savefig.dpi": 300, "figure.dpi": 100,
    "font.family": "DejaVu Sans",  # deterministic font
})
EDGE_COL, EDGE_W = "#58595B", 0.5
pt_to_in = lambda pt: pt / 72.0

def _style_axes(ax, grid_y=True):
    # ticks
    ax.tick_params(axis="y", which="major", width=EDGE_W, length=6, color=EDGE_COL)
    ax.tick_params(axis="y", which="minor", width=EDGE_W, length=3, color=EDGE_COL)
    ax.tick_params(axis="x", which="both",  width=EDGE_W, length=3, color=EDGE_COL)
    # spines
    for s in ["left", "bottom"]:
        ax.spines[s].set_linewidth(EDGE_W)
        ax.spines[s].set_color(EDGE_COL)
    for s in ["top", "right"]:
        ax.spines[s].set_visible(False)
    # grid
    if grid_y:
        ax.grid(True, axis="y", which="major", linewidth=0.3, alpha=0.3)

# ======= Inputs you already have =======
# hex_ids: List[str] -> e.g., ["0x0239", "0x040B", ...]

# ======= Path templates (EDITED SA path as before) =======
SA_ROOT_TEMPLATE = (
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/logic_synthesis_SA/11400_steps/4_inputs/no_fw_preference/{HEX}/seed_1"
)

FINETUNE_ROOT_TEMPLATE = (
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/fine_tune/GAT_MLP_with_scalars/4000nn/initial_state_sampling_factor_0/"
    "4_inputs/{HEX}/seed_1"
)

TB_EVENT_GLOBS = ["**/events.out.tfevents.*", "**/*tfevents*"]
LOG_GLOBS = ["**/slurm*.log", "slurm*.log"]  # search root + subdirs

TB_TAG = "custom/best_energy"
ASSUME_TAG_IS_BEST = True  # True if tag is already "best so far"

# ---------------- Helpers ----------------
def _norm_hex(x: str) -> str:
    s = str(x).strip().lower()
    if not s.startswith("0x"): s = "0x" + s
    return "0x" + s[2:].upper()

def sa_root_for_hex(hex_id: str) -> Path:
    return Path(SA_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

def ft_root_for_hex(hex_id: str) -> Path:
    return Path(FINETUNE_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

# ---- SA loaders ----
def find_sa_csv(root: Path) -> Optional[Path]:
    p = root / "sa_masked" / "sa_step_metrics.csv"
    if p.exists(): return p
    for pat in ["**/sa_step_metrics.csv", "**/*sa_step_metrics.csv"]:
        hits = list(root.glob(pat))
        if hits: return sorted(hits)[0]
    return None

def sa_final_best(csv_path: Path) -> Optional[float]:
    try:
        df = pd.read_csv(csv_path)
        if "step" in df.columns:
            df["step"] = pd.to_numeric(df["step"], errors="coerce")
            df = df.dropna(subset=["step"]).sort_values("step", kind="mergesort")
        if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
            s = pd.to_numeric(df["best_energy_so_far"], errors="coerce").dropna()
            if len(s): return float(s.iloc[-1])
        if "energy" in df.columns and df["energy"].notna().any():
            e = pd.to_numeric(df["energy"], errors="coerce").dropna()
            if len(e): return float(e.cummin().iloc[-1])
    except Exception as e:
        print(f"[WARN] SA read failed {csv_path}: {e}")
    return None

def load_sa_series(csv_path: Path) -> Tuple[np.ndarray, np.ndarray]:
    """
    Returns (sa_steps, sa_best_so_far) as numpy arrays.
    Uses 'best_energy_so_far' if present; else cummin('energy').
    """
    try:
        df = pd.read_csv(csv_path)
    except Exception as e:
        print(f"[WARN] SA CSV read failed {csv_path}: {e}")
        return np.array([], dtype=int), np.array([], dtype=float)
    if "step" not in df.columns:
        return np.array([], dtype=int), np.array([], dtype=float)
    df["step"] = pd.to_numeric(df["step"], errors="coerce")
    df = df.dropna(subset=["step"]).sort_values("step", kind="mergesort")
    steps = df["step"].astype(int).to_numpy()
    if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
        vals = pd.to_numeric(df["best_energy_so_far"], errors="coerce").to_numpy()
        # enforce monotone nonincreasing (lower is better)
        vals = np.minimum.accumulate(vals)
    elif "energy" in df.columns and df["energy"].notna().any():
        eng = pd.to_numeric(df["energy"], errors="coerce").to_numpy()
        vals = np.minimum.accumulate(eng)
    else:
        return np.array([], dtype=int), np.array([], dtype=float)
    return steps, vals.astype(float)

# ---- Fine-tune loaders (TensorBoard) ----
def find_event_runs(root: Path) -> List[Path]:
    files = []
    for pat in TB_EVENT_GLOBS:
        files.extend(root.glob(pat))
    return sorted(set(p.parent for p in files), key=lambda p: p.as_posix())

def load_tb_scalar(run_dir: Path, tag: str) -> Tuple[np.ndarray, np.ndarray]:
    try:
        ea = event_accumulator.EventAccumulator(str(run_dir))
        ea.Reload()
        if tag not in ea.Tags().get("scalars", []):
            return np.array([], dtype=int), np.array([], dtype=float)
        ev = ea.Scalars(tag)
        steps  = np.fromiter((e.step for e in ev), dtype=int)
        values = np.fromiter((e.value for e in ev), dtype=float)
        return steps, values
    except Exception as e:
        print(f"[WARN] TB load failed in {run_dir}: {e}")
        return np.array([], dtype=int), np.array([], dtype=float)

def finetune_final_best(root: Path, tag: str, assume_best_series=True) -> Tuple[Optional[float], Optional[Path]]:
    best_val, best_dir = None, None
    for d in find_event_runs(root):
        steps, vals = load_tb_scalar(d, tag)
        if steps.size == 0: continue
        final = float(vals[-1]) if assume_best_series else float(np.nanmin(vals))
        if best_val is None or final < best_val:
            best_val, best_dir = final, d
    return best_val, best_dir

# ---- Log discovery + parsing ----
def _find_latest_log(root: Path) -> Optional[Path]:
    hits = []
    for pat in LOG_GLOBS:
        hits.extend(root.glob(pat))
    if not hits: return None
    hits = sorted(hits, key=lambda p: p.stat().st_mtime, reverse=True)
    return hits[0]

_sa_re = re.compile(r"Total time\s*=\s*([\d\.]+)s")
_ft_re = re.compile(r"Completed in\s*(\d+)h\s+(\d+)m\s+(\d+)s", re.IGNORECASE)

def parse_sa_time_seconds(log_path: Path) -> Optional[float]:
    try:
        txt = log_path.read_text(errors="ignore")
        m = None
        for m in _sa_re.finditer(txt):
            pass  # take last match
        if m:
            return float(m.group(1))
    except Exception as e:
        print(f"[WARN] SA time parse failed {log_path}: {e}")
    return None

def parse_ft_time_seconds(log_path: Path) -> Optional[float]:
    try:
        txt = log_path.read_text(errors="ignore")
        m = None
        for m in _ft_re.finditer(txt):
            pass  # take last match
        if m:
            h, mnt, s = map(int, m.groups())
            return float(h)*3600 + float(mnt)*60 + float(s)
    except Exception as e:
        print(f"[WARN] FT time parse failed {log_path}: {e}")
    return None

def _best_so_far(vals: np.ndarray) -> np.ndarray:
    if vals.size == 0: 
        return vals
    return np.minimum.accumulate(vals)

# step where the (first) global minimum is reached
def _step_of_first_global_min(steps: Optional[np.ndarray], vals: Optional[np.ndarray]) -> float:
    if steps is None or vals is None or len(steps) == 0 or len(vals) == 0:
        return float("nan")
    idx = int(np.nanargmin(vals))
    return float(steps[idx])

def compare_sa_vs_finetune_with_time(
    hex_ids: List[str],
    tag: str = TB_TAG,
    assume_tag_is_best: bool = ASSUME_TAG_IS_BEST,
    make_plots: bool = True,
    show_plots: bool = True,
    save_quality_bar_path: Optional[Path] = None,
    save_quality_scatter_path: Optional[Path] = None,
    save_speed_bar_path: Optional[Path] = None,
    save_time_scatter_path: Optional[Path] = None,
    save_speedup_plot_path: Optional[Path] = None,
    save_summary_csv: Optional[Path] = None,
    # matched-steps % figure:
    save_pct_over_steps_path: Optional[Path] = None,
    step_bin: int = 1000,
    # NEW: Ties  steps to best
    make_tie_steps_plot: bool = True,
    save_tie_steps_plot_path: Optional[Path] = None,
):
    """
    Per hex:
      - SA_best := final best from sa_step_metrics.csv
      - SA_series := (step, best_so_far)
      - FT_best := final (or min) from TB tag
      - FT_series := (step, best_so_far) from TB scalars
      - SA_time / FT_time from logs (optional)

    Plots:
      1) Quality bar
      2) Quality scatter
      3) Speed bar (if both times)
      4) Time scatter (log-log) (if both times)
      5) Per-circuit speedup bars (if both times)
      6) Percentages over matched training steps (bins of `step_bin`)
      7) NEW: Ties  steps to reach best energy (global min), SA vs FT
    """

    def _best_so_far(vals: np.ndarray) -> np.ndarray:
        if vals is None or vals.size == 0:
            return vals
        return np.minimum.accumulate(vals)

    def _step_of_first_global_min(steps: Optional[np.ndarray], vals: Optional[np.ndarray]) -> float:
        """Return the step where the (first) global minimum is reached, or NaN."""
        if steps is None or vals is None or len(steps) == 0 or len(vals) == 0:
            return float("nan")
        idx = int(np.nanargmin(vals))
        return float(steps[idx])

    rows = []
    missing = []

    # Track counts + exact hex IDs per bucket
    quality_counts = {"SA better": 0, "FT better": 0, "Tie": 0}
    sa_better_hex, ft_better_hex, tie_hex = [], [], []

    for hex_id in hex_ids:
        hex_norm = _norm_hex(hex_id)

        # SA side
        sa_root = sa_root_for_hex(hex_norm)
        sa_csv  = find_sa_csv(sa_root)
        sa_best = sa_final_best(sa_csv) if sa_csv else None
        sa_steps, sa_vals = load_sa_series(sa_csv) if sa_csv else (np.array([], int), np.array([], float))
        sa_log  = _find_latest_log(sa_root)
        sa_time = parse_sa_time_seconds(sa_log) if sa_log else None

        # FT side
        ft_root = ft_root_for_hex(hex_norm)
        ft_best, ft_dir = finetune_final_best(ft_root, tag, assume_tag_is_best)
        ft_steps, ft_vals = (np.array([], int), np.array([], float))
        if ft_dir:
            ft_steps, ft_vals = load_tb_scalar(ft_dir, tag)
            if ft_steps.size > 0:
                order = np.argsort(ft_steps, kind="mergesort")
                ft_steps = ft_steps[order].astype(int)
                ft_vals  = ft_vals[order].astype(float)

        # Enforce best-so-far series (robustness)
        if sa_vals.size:
            sa_vals = _best_so_far(sa_vals.astype(float))
        if ft_vals.size:
            if not assume_tag_is_best:
                ft_vals = _best_so_far(ft_vals)
            else:
                ft_vals = _best_so_far(ft_vals)

        ft_log  = _find_latest_log(ft_root)
        ft_time = parse_ft_time_seconds(ft_log) if ft_log else None

        if sa_best is None or ft_best is None:
            missing.append(hex_norm)
            continue

        # Quality verdict (finals)
        if sa_best < ft_best:
            q_verdict = "SA better"
            quality_counts["SA better"] += 1
            sa_better_hex.append(hex_norm)
        elif sa_best > ft_best:
            q_verdict = "FT better"
            quality_counts["FT better"] += 1
            ft_better_hex.append(hex_norm)
        else:
            q_verdict = "Tie"
            quality_counts["Tie"] += 1
            tie_hex.append(hex_norm)

        # Speed verdict & folds
        if (sa_time is not None) and (ft_time is not None):
            if sa_time < ft_time:
                s_verdict = "SA faster"
            elif sa_time > ft_time:
                s_verdict = "FT faster"
            else:
                s_verdict = "Tie"
            fc_ft_over_sa = float(ft_time) / float(sa_time) if sa_time and sa_time > 0 else np.nan
            speedup_ft_vs_sa = float(sa_time) / float(ft_time) if ft_time and ft_time > 0 else np.nan
        else:
            s_verdict = "N/A"
            fc_ft_over_sa = np.nan
            speedup_ft_vs_sa = np.nan

        # Steps where the (first) global minima were achieved
        sa_step_best = _step_of_first_global_min(sa_steps, sa_vals)
        ft_step_best = _step_of_first_global_min(ft_steps, ft_vals)

        rows.append({
            "hex_id": hex_norm,
            "SA_best": float(sa_best),
            "FT_best": float(ft_best),
            "quality_result": q_verdict,
            "SA_time_s": sa_time,
            "FT_time_s": ft_time,
            "speed_result": s_verdict,
            "time_fold_change_FT_over_SA": fc_ft_over_sa,      # >1  FT slower than SA
            "speedup_FT_vs_SA": speedup_ft_vs_sa,              # >1  FT faster than SA
            "ft_chosen_run": str(ft_dir) if ft_dir else "",
            "sa_metrics_csv": str(sa_csv) if sa_csv else "",
            "sa_log": str(sa_log) if sa_log else "",
            "ft_log": str(ft_log) if ft_log else "",
            # stash step series for matched-steps figure
            "_sa_steps": sa_steps, "_sa_vals": sa_vals,
            "_ft_steps": ft_steps, "_ft_vals": ft_vals,
            # NEW: persisted for plotting/CSV
            "SA_steps_to_best": sa_step_best,
            "FT_steps_to_best": ft_step_best,
        })

    df = pd.DataFrame(rows)
    n_used = len(df)
    print(f"[SA vs FT] quality counts: {quality_counts} out of N={n_used} (missing {len(missing)})")

    # Print the exact circuits per bucket
    if sa_better_hex:
        print(f"  SA wins   ({len(sa_better_hex)}): {', '.join(sorted(sa_better_hex))}")
    if ft_better_hex:
        print(f"  FT wins   ({len(ft_better_hex)}): {', '.join(sorted(ft_better_hex))}")
    if tie_hex:
        print(f"  Ties      ({len(tie_hex)}): {', '.join(sorted(tie_hex))}")
    if missing:
        print(f"  Missing   ({len(missing)}): {', '.join(sorted(missing))}")

    # Speed summary over circuits with both times
    both_times = df.dropna(subset=["SA_time_s", "FT_time_s"])
    speed_den = len(both_times)
    if speed_den > 0:
        speed_summary = {
            "SA_faster_%": 100.0 * (both_times["speed_result"] == "SA faster").sum() / speed_den,
            "FT_faster_%": 100.0 * (both_times["speed_result"] == "FT faster").sum() / speed_den,
            "Tie_%":       100.0 * (both_times["speed_result"] == "Tie").sum()       / speed_den,
            "Denominator": speed_den,
        }
    else:
        speed_summary = {"SA_faster_%": 0.0, "FT_faster_%": 0.0, "Tie_%": 0.0, "Denominator": 0}
    print(f"[SA vs FT] speed counts: SA faster={(both_times['speed_result']=='SA faster').sum()} | "
          f"FT faster={(both_times['speed_result']=='FT faster').sum()} | "
          f"Tie={(both_times['speed_result']=='Tie').sum()}  over {speed_den} circuits with both times")

    # Summary
    summary = pd.DataFrame([{
        "N_used": n_used,
        "SA_better": quality_counts["SA better"],
        "FT_better": quality_counts["FT better"],
        "Tie": quality_counts["Tie"],
        "SA_better_%": 100.0 * quality_counts["SA better"] / n_used if n_used else 0.0,
        "FT_better_%": 100.0 * quality_counts["FT better"] / n_used if n_used else 0.0,
        "Tie_%": 100.0 * quality_counts["Tie"] / n_used if n_used else 0.0,
        "Speed_denominator": speed_den,
        "Speed_SA_faster_%": speed_summary["SA_faster_%"],
        "Speed_FT_faster_%": speed_summary["FT_faster_%"],
        "Speed_Tie_%": speed_summary["Tie_%"],
        "N_missing": len(missing),
        "Missing_hex_ids": ",".join(sorted(missing)),
        "SA_better_hex_ids": ",".join(sorted(sa_better_hex)),
        "FT_better_hex_ids": ",".join(sorted(ft_better_hex)),
        "Tie_hex_ids": ",".join(sorted(tie_hex)),
        "tb_tag": tag,
        "assume_tag_is_best": assume_tag_is_best,
        "median_time_fold_change_FT_over_SA": np.nanmedian(df["time_fold_change_FT_over_SA"]) if not df.empty else np.nan,
        "median_speedup_FT_vs_SA": np.nanmedian(df["speedup_FT_vs_SA"]) if not df.empty else np.nan,
    }])

    # ---- Plots (styled, half-size canvases) ----
    figs = {}
    if make_plots:
        # 1) Quality bar
        cats_q = ["SA better", "FT better", "Tie"]
        vals_q = [
            summary["SA_better_%"].iloc[0],
            summary["FT_better_%"].iloc[0],
            summary["Tie_%"].iloc[0],
        ]
        fig_qbar, ax = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
        ax.bar(cats_q, vals_q, color=["#4CAF50", "#F44336", "#9E9E9E"],
               edgecolor=EDGE_COL, linewidth=EDGE_W)
        ax.set_ylim(0, 100)
        ax.set_ylabel("Percentage of circuits")
        ax.set_title(f"SA vs Fine-tuning  Quality (N={n_used})")
        for i, v in enumerate(vals_q):
            ax.text(i, min(100, v) + 0.5, f"{v:.1f}%", ha="center", va="bottom")
        _style_axes(ax, grid_y=True)
        fig_qbar.tight_layout()
        figs["quality_bar"] = fig_qbar
        if save_quality_bar_path:
            Path(save_quality_bar_path).parent.mkdir(parents=True, exist_ok=True)
            fig_qbar.savefig(save_quality_bar_path, bbox_inches="tight")
        if show_plots:
            plt.show()

        # 2) Quality scatter
        if not df.empty:
            fig_qs, ax2 = plt.subplots(figsize=(pt_to_in(100), pt_to_in(87.5)), dpi=300)
            ax2.scatter(df["SA_best"], df["FT_best"], s=8,
                        edgecolors=EDGE_COL, linewidths=EDGE_W/2)
            lo = float(min(df["SA_best"].min(), df["FT_best"].min()))
            hi = float(max(df["SA_best"].max(), df["FT_best"].max()))
            ax2.plot([lo, hi], [lo, hi], "--", lw=EDGE_W, color=EDGE_COL)
            ax2.set_xlabel("SA best (lower is better)")
            ax2.set_ylabel("Fine-tune best (lower is better)")
            ax2.set_title("Per-circuit quality")
            ax2.set_aspect("equal", adjustable="box")
            _style_axes(ax2, grid_y=True)
            fig_qs.tight_layout()
            figs["quality_scatter"] = fig_qs
            if save_quality_scatter_path:
                Path(save_quality_scatter_path).parent.mkdir(parents=True, exist_ok=True)
                fig_qs.savefig(save_quality_scatter_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # --- 2b) NEW: Ties  steps to reach best energy (global min)
        if make_tie_steps_plot and not df.empty:
            ties_only = df[df["quality_result"] == "Tie"].copy()
            ties_only = ties_only.dropna(subset=["SA_steps_to_best", "FT_steps_to_best"])
            if not ties_only.empty:
                fig_tie, ax_t = plt.subplots(figsize=(pt_to_in(100), pt_to_in(87.5)), dpi=300)
                x_sa = ties_only["SA_steps_to_best"].astype(float).to_numpy()
                y_ft = ties_only["FT_steps_to_best"].astype(float).to_numpy()
                ax_t.scatter(x_sa, y_ft, s=8, edgecolors=EDGE_COL, linewidths=EDGE_W/2)

                lo = float(min(np.nanmin(x_sa), np.nanmin(y_ft)))
                hi = float(max(np.nanmax(x_sa), np.nanmax(y_ft)))
                pad = 0.02 * (hi - lo) if hi > lo else 1.0
                ax_t.plot([lo - pad, hi + pad], [lo - pad, hi + pad], "--", lw=EDGE_W, color=EDGE_COL)

                ax_t.set_xlabel("SA steps to best (lower is faster)")
                ax_t.set_ylabel("Fine-tune steps to best (lower is faster)")
                ax_t.set_title(f"Ties: Steps to Reach Best Energy (N={len(ties_only)})")
                ax_t.set_aspect("equal", adjustable="box")
                _style_axes(ax_t, grid_y=True)
                fig_tie.tight_layout()
                figs["tie_steps_scatter"] = fig_tie

                if save_tie_steps_plot_path:
                    Path(save_tie_steps_plot_path).parent.mkdir(parents=True, exist_ok=True)
                    fig_tie.savefig(save_tie_steps_plot_path, bbox_inches="tight")
                if show_plots:
                    plt.show()
            else:
                print("[Info] No quality ties with step info; skipping tie-steps plot.")

        # 3) Speed bar (if times)
        df_times = df.dropna(subset=["SA_time_s", "FT_time_s"])
        speed_den = len(df_times)
        if speed_den > 0:
            cats_s = ["SA faster", "FT faster", "Tie"]
            vals_s = [
                100.0 * (df_times["speed_result"] == "SA faster").sum() / speed_den,
                100.0 * (df_times["speed_result"] == "FT faster").sum() / speed_den,
                100.0 * (df_times["speed_result"] == "Tie").sum() / speed_den,
            ]
            fig_sbar, ax3 = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
            ax3.bar(cats_s, vals_s, color=["#1976D2", "#E65100", "#9E9E9E"],
                    edgecolor=EDGE_COL, linewidth=EDGE_W)
            ax3.set_ylim(0, 100)
            ax3.set_ylabel(f"Percentage of circuits (N={speed_den})")
            ax3.set_title("Speed comparison")
            for i, v in enumerate(vals_s):
                ax3.text(i, min(100, v) + 0.5, f"{v:.1f}%", ha="center", va="bottom")
            _style_axes(ax3, grid_y=True)
            fig_sbar.tight_layout()
            figs["speed_bar"] = fig_sbar
            if save_speed_bar_path:
                Path(save_speed_bar_path).parent.mkdir(parents=True, exist_ok=True)
                fig_sbar.savefig(save_speed_bar_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # 4) Time scatter (log-log), FT_time (x) vs SA_time (y)
        if not df_times.empty:
            fig_tscatter, ax4 = plt.subplots(figsize=(pt_to_in(100), pt_to_in(87.5)), dpi=300)
            x = df_times["FT_time_s"].astype(float).to_numpy()
            y = df_times["SA_time_s"].astype(float).to_numpy()
            ax4.scatter(x, y, s=8, edgecolors=EDGE_COL, linewidths=EDGE_W/2)
            lo = float(min(np.min(x), np.min(y))); hi = float(max(np.max(x), np.max(y)))
            lo = max(lo, 1e-3)
            ax4.plot([lo, hi], [lo, hi], "--", lw=EDGE_W, color=EDGE_COL)
            ax4.set_xscale("log"); ax4.set_yscale("log")
            ax4.set_xlabel("Fine-tune wall time (s)")
            ax4.set_ylabel("SA wall time (s)")
            ax4.set_title("Per-circuit time (log-log)")
            _style_axes(ax4, grid_y=True)
            fig_tscatter.tight_layout()
            figs["time_scatter"] = fig_tscatter
            if save_time_scatter_path:
                Path(save_time_scatter_path).parent.mkdir(parents=True, exist_ok=True)
                fig_tscatter.savefig(save_time_scatter_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # 5) Per-circuit speedup plot (FT vs SA)  highlight QUALITY ties
        df_speedup = df_times.copy()
        if not df_speedup.empty:
            df_speedup["speedup_FT_vs_SA"] = df_speedup["speedup_FT_vs_SA"].astype(float)
            df_speedup = df_speedup.dropna(subset=["speedup_FT_vs_SA"])
        if not df_speedup.empty:
            from matplotlib.patches import Patch  # local import to avoid top-level dependency if unused
            is_quality_tie = df_speedup["quality_result"].eq("Tie").to_numpy()
            base_color = "#CDCDCD"
            tie_color  = "#1ADCFA"
            bar_colors = [tie_color if t else base_color for t in is_quality_tie]
            fig_fc, ax5 = plt.subplots(figsize=(pt_to_in(max(150, 6 * len(df_speedup))), pt_to_in(90)), dpi=300)
            ax5.bar(df_speedup["hex_id"], df_speedup["speedup_FT_vs_SA"],
                    color=bar_colors, edgecolor=EDGE_COL, linewidth=EDGE_W)
            ax5.axhline(1.0, color=EDGE_COL, ls="--", lw=EDGE_W)
            ax5.set_ylabel("Speedup (FT vs SA)  =  SA_time / FT_time")
            ax5.set_xlabel("Circuit (hex)")
            ax5.set_title("Per-circuit speedup (>1  fine-tune faster)")
            ymax = float(df_speedup["speedup_FT_vs_SA"].max()) * 1.1
            ax5.set_ylim(0, max(1.05, ymax))
            ax5.set_xticklabels(df_speedup["hex_id"], rotation=60, ha="right")
            ax5.legend(handles=[
                Patch(facecolor=base_color, edgecolor=EDGE_COL, linewidth=EDGE_W, label="Non-tie (quality)"),
                Patch(facecolor=tie_color,  edgecolor=EDGE_COL, linewidth=EDGE_W, label="Tie (quality)"),
            ], frameon=False)
            _style_axes(ax5, grid_y=True)
            highlighted = df_speedup.loc[is_quality_tie, "hex_id"].tolist()
            print(f"[plot] Quality ties highlighted in speedup plot: {highlighted}")
            fig_fc.tight_layout()
            figs["speedup_plot"] = fig_fc
            if save_speedup_plot_path:
                Path(save_speedup_plot_path).parent.mkdir(parents=True, exist_ok=True)
                fig_fc.savefig(save_speedup_plot_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # 6) Percentages over matched steps (bins of `step_bin`)
        if not df.empty:
            series_by_hex = {}
            global_max_bin = 0
            for _, r in df.iterrows():
                sa_s, sa_v = r["_sa_steps"], r["_sa_vals"]
                ft_s, ft_v = r["_ft_steps"], r["_ft_vals"]
                if sa_s is None or ft_s is None: 
                    continue
                if len(sa_s) == 0 or len(ft_s) == 0:
                    continue
                # ensure monotone steps + best-so-far
                sa_order = np.argsort(sa_s, kind="mergesort")
                ft_order = np.argsort(ft_s, kind="mergesort")
                sa_s = sa_s[sa_order].astype(int); sa_v = _best_so_far(np.asarray(sa_v)[sa_order].astype(float))
                ft_s = ft_s[ft_order].astype(int); ft_v = _best_so_far(np.asarray(ft_v)[ft_order].astype(float))
                series_by_hex[r["hex_id"]] = (sa_s, sa_v, ft_s, ft_v)
                # upper bound per circuit = min(max steps)
                candidate_max = min(int(sa_s[-1]), int(ft_s[-1]))
                if candidate_max > global_max_bin:
                    global_max_bin = candidate_max

            if series_by_hex and global_max_bin >= step_bin:
                bins = np.arange(step_bin, global_max_bin + step_bin, step_bin, dtype=int)

                ft_better_pct, sa_better_pct, tie_pct, den = [], [], [], []
                for b in bins:
                    ft_better = sa_better = tie = denom = 0
                    for _, (sa_s, sa_v, ft_s, ft_v) in series_by_hex.items():
                        # last index with step <= b
                        i_sa = np.searchsorted(sa_s, b, side="right") - 1
                        i_ft = np.searchsorted(ft_s, b, side="right") - 1
                        if i_sa < 0 or i_ft < 0:
                            continue
                        denom += 1
                        sa_val = float(sa_v[i_sa])
                        ft_val = float(ft_v[i_ft])
                        if ft_val < sa_val:
                            ft_better += 1
                        elif ft_val > sa_val:
                            sa_better += 1
                        else:
                            tie += 1
                    den.append(denom)
                    if denom > 0:
                        ft_better_pct.append(100.0 * ft_better / denom)
                        sa_better_pct.append(100.0 * sa_better / denom)
                        tie_pct.append(100.0 * tie / denom)
                    else:
                        ft_better_pct.append(0.0)
                        sa_better_pct.append(0.0)
                        tie_pct.append(0.0)

                fig_pct, ax6 = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
                ax6.plot(bins, ft_better_pct, label="FT better %")
                ax6.plot(bins, sa_better_pct, label="SA better %")
                ax6.plot(bins, tie_pct,       label="Tie %")
                ax6.set_ylim(0, 100)
                ax6.set_xlabel(f"Training steps (bins of {step_bin})")
                ax6.set_ylabel("Percentage of circuits")
                last_den = den[-1] if len(den) else 0
                ax6.set_title(f"Matched-steps comparison (N up to {last_den} per bin)")
                _style_axes(ax6, grid_y=True)
                ax6.legend(frameon=False)
                fig_pct.tight_layout()
                figs["percentages_over_matched_steps"] = fig_pct
                if save_pct_over_steps_path:
                    Path(save_pct_over_steps_path).parent.mkdir(parents=True, exist_ok=True)
                    fig_pct.savefig(save_pct_over_steps_path, bbox_inches="tight")
                if show_plots:
                    plt.show()

    # Save CSVs
    if save_summary_csv:
        save_summary_csv = Path(save_summary_csv)
        save_summary_csv.parent.mkdir(parents=True, exist_ok=True)
        summary.to_csv(save_summary_csv, index=False)
        # drop the series columns before writing per-hex table
        df_to_save = df.drop(columns=[c for c in ["_sa_steps","_sa_vals","_ft_steps","_ft_vals"] if c in df.columns])
        df_to_save.to_csv(save_summary_csv.with_name(save_summary_csv.stem + "_per_hex.csv"), index=False)

    return summary, df, figs


summary_df, per_hex_df, figs = compare_sa_vs_finetune_with_time(
    hex_ids,
    tag="custom/best_energy",
    assume_tag_is_best=True,
    make_plots=True,
    show_plots=True
)
```

```python
# === Trained (registry sampling) vs Simulated Annealing (SA) ===
# - Trained metrics: <TRAINED_ROOT>/trained_masked/*trained_episode_metrics.csv  (or .../trajectories/*episode_metrics.csv)
# - SA metrics:      <SA_ROOT>/sa_masked/sa_step_metrics.csv
# - Logs (optional): slurm*.log with either:
#       SA: "[SA] Finished ... Total time=4175.0s"
#       Generic: "Completed in 0h 14m 2s"  (works for trained too if present)

import re
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib as mpl
from typing import List, Optional, Tuple, Dict
from matplotlib.patches import Patch

# ---------- unified export-friendly style (5 pt) ----------
mpl.rcParams.update({
    "font.size": 5, "axes.titlesize": 5, "axes.labelsize": 5,
    "legend.fontsize": 5, "xtick.labelsize": 5, "ytick.labelsize": 5,
    "pdf.fonttype": 42, "ps.fonttype": 42, "pdf.use14corefonts": False, "svg.fonttype": "none",
    "axes.linewidth": 0.5, "lines.linewidth": 0.5,
    "legend.frameon": False, "savefig.dpi": 300, "figure.dpi": 100,
    "font.family": "DejaVu Sans",
})
EDGE_COL, EDGE_W = "#58595B", 0.5
pt_to_in = lambda pt: pt / 72.0

def _style_axes(ax, grid_y=True):
    ax.tick_params(axis="y", which="major", width=EDGE_W, length=6, color=EDGE_COL)
    ax.tick_params(axis="y", which="minor", width=EDGE_W, length=3, color=EDGE_COL)
    ax.tick_params(axis="x", which="both",  width=EDGE_W, length=3, color=EDGE_COL)
    for s in ["left", "bottom"]:
        ax.spines[s].set_linewidth(EDGE_W)
        ax.spines[s].set_color(EDGE_COL)
    for s in ["top", "right"]:
        ax.spines[s].set_visible(False)
    if grid_y:
        ax.grid(True, axis="y", which="major", linewidth=0.3, alpha=0.3)

# ---------- path templates (edit TRAINED if your layout differs) ----------
TRAINED_ROOT_TEMPLATE = (    
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/manuscript/"
    "zero-shot/against_random/GAT_MLP_with_scalars/initial_state_sampling_factor_0/"
    "4_inputs_{HEX}_samplingfactor_0_4000nn/seed_1"       
)

# no fw preference

SA_ROOT_TEMPLATE = (
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/logic_synthesis_SA/1000_steps/4_inputs/no_fw_preference/{HEX}/seed_1"
)


LOG_GLOBS = ["**/slurm*.log", "slurm*.log"]

def _norm_hex(x: str) -> str:
    s = str(x).strip().lower()
    if not s.startswith("0x"): s = "0x" + s
    return "0x" + s[2:].upper()

def trained_root_for_hex(hex_id: str) -> Path:
    return Path(TRAINED_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

def sa_root_for_hex(hex_id: str) -> Path:
    return Path(SA_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

# ---------- trained metrics discovery ----------
TRAINED_CSV_PATTERNS = [
    "trained_masked/*trained_episode_metrics.csv",
    "trained_masked/trajectories/*episode_metrics.csv",
    "**/*trained_episode_metrics.csv",
    "**/trajectories/*episode_metrics.csv",
]

def find_trained_csv(root: Path) -> Optional[Path]:
    for pat in TRAINED_CSV_PATTERNS:
        hits = sorted(set(p for p in root.glob(pat)))
        if hits:
            return hits[0]
    for name in ("trained_episode_metrics.csv",):
        p = root / name
        if p.exists(): return p
    return None

def trained_final_best(csv_path: Path) -> Optional[float]:
    try:
        df = pd.read_csv(csv_path)
        if "episode" in df.columns:
            df = df.sort_values("episode", kind="mergesort")
        if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
            s = pd.to_numeric(df["best_energy_so_far"], errors="coerce").dropna()
            if len(s): return float(s.iloc[-1])
        if "best_energy_in_episode" in df.columns and df["best_energy_in_episode"].notna().any():
            s = pd.to_numeric(df["best_energy_in_episode"], errors="coerce")
            cm = s.cummin().dropna()
            if len(cm): return float(cm.iloc[-1])
    except Exception as e:
        print(f"[WARN] trained read failed {csv_path}: {e}")
    return None

# ---------- SA metrics discovery ----------
def find_sa_csv(root: Path) -> Optional[Path]:
    p = root / "sa_masked" / "sa_step_metrics.csv"
    if p.exists(): return p
    for pat in ["**/sa_step_metrics.csv", "**/*sa_step_metrics.csv"]:
        hits = sorted(set(p for p in root.glob(pat)))
        if hits: return hits[0]
    return None

def sa_final_best(csv_path: Path) -> Optional[float]:
    try:
        df = pd.read_csv(csv_path)
        if "step" in df.columns:
            df["step"] = pd.to_numeric(df["step"], errors="coerce")
            df = df.dropna(subset=["step"]).sort_values("step", kind="mergesort")
        if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
            s = pd.to_numeric(df["best_energy_so_far"], errors="coerce").dropna()
            if len(s): return float(s.iloc[-1])
        if "energy" in df.columns and df["energy"].notna().any():
            e = pd.to_numeric(df["energy"], errors="coerce").dropna()
            if len(e): return float(e.cummin().iloc[-1])
    except Exception as e:
        print(f"[WARN] SA read failed {csv_path}: {e}")
    return None

# ---------- log parsing ----------
_sa_total_re = re.compile(r"Total time\s*=\s*([\d\.]+)s")
_generic_hms_re = re.compile(r"Completed in\s*(\d+)h\s+(\d+)m\s+(\d+)s", re.IGNORECASE)

def _find_latest_log(root: Path) -> Optional[Path]:
    hits = []
    for pat in LOG_GLOBS:
        hits.extend(root.glob(pat))
    if not hits: return None
    hits = sorted(hits, key=lambda p: p.stat().st_mtime, reverse=True)
    return hits[0]

def parse_sa_time_seconds(log_path: Path) -> Optional[float]:
    try:
        txt = log_path.read_text(errors="ignore")
        m = None
        for m in _sa_total_re.finditer(txt):
            pass
        if m:
            return float(m.group(1))
        m2 = None
        for m2 in _generic_hms_re.finditer(txt):
            pass
        if m2:
            h, m, s = map(int, m2.groups())
            return h*3600 + m*60 + s
    except Exception as e:
        print(f"[WARN] SA time parse failed {log_path}: {e}")
    return None

def parse_trained_time_seconds(log_path: Path) -> Optional[float]:
    try:
        txt = log_path.read_text(errors="ignore")
        m = None
        for m in _generic_hms_re.finditer(txt):
            pass
        if m:
            h, mnt, s = map(int, m.groups())
            return h*3600 + mnt*60 + s
        m2 = None
        for m2 in _sa_total_re.finditer(txt):
            pass
        if m2:
            return float(m2.group(1))
    except Exception as e:
        print(f"[WARN] trained time parse failed {log_path}: {e}")
    return None

# ---------- main comparer ----------
def compare_trained_vs_sa(
    hex_ids: List[str],
    make_plots: bool = True,
    show_plots: bool = True,
    save_quality_bar_path: Optional[Path] = None,
    save_quality_scatter_path: Optional[Path] = None,
    save_speed_bar_path: Optional[Path] = None,
    save_time_scatter_path: Optional[Path] = None,
    save_speedup_plot_path: Optional[Path] = None,
    save_summary_csv: Optional[Path] = None,
    # NEW:
    print_summaries: bool = True,
):
    """
    Quality: lower energy is better.
    Speed: lower wall time (s) is faster.
    Speedup (trained vs SA): SA_time / trained_time  (>1  trained faster).
    """
    rows = []
    missing = []

    quality_counts = {"Trained better": 0, "SA better": 0, "Tie": 0}
    trained_better_hex, sa_better_hex, tie_hex = [], [], []

    speed_counts = {"Trained faster": 0, "SA faster": 0, "Tie": 0}
    speed_den = 0

    for hex_id in hex_ids:
        hex_norm = _norm_hex(hex_id)
        tr_root = trained_root_for_hex(hex_norm)
        sa_root = sa_root_for_hex(hex_norm)

        tr_csv = find_trained_csv(tr_root)
        sa_csv = find_sa_csv(sa_root)
        tr_best = trained_final_best(tr_csv) if tr_csv else None
        sa_best = sa_final_best(sa_csv) if sa_csv else None

        tr_log = _find_latest_log(tr_root)
        sa_log = _find_latest_log(sa_root)
        tr_time = parse_trained_time_seconds(tr_log) if tr_log else None
        sa_time = parse_sa_time_seconds(sa_log) if sa_log else None

        if tr_best is None or sa_best is None:
            missing.append(hex_norm)
            continue

        if tr_best < sa_best:
            qv = "Trained better"; quality_counts["Trained better"] += 1; trained_better_hex.append(hex_norm)
        elif tr_best > sa_best:
            qv = "SA better";      quality_counts["SA better"] += 1;      sa_better_hex.append(hex_norm)
        else:
            qv = "Tie";            quality_counts["Tie"] += 1;            tie_hex.append(hex_norm)

        if (tr_time is not None) and (sa_time is not None):
            speed_den += 1
            if tr_time < sa_time:
                sv = "Trained faster"; speed_counts["Trained faster"] += 1
            elif tr_time > sa_time:
                sv = "SA faster";      speed_counts["SA faster"] += 1
            else:
                sv = "Tie";            speed_counts["Tie"] += 1
            speedup_trained_vs_sa = float(sa_time) / float(tr_time) if tr_time and tr_time > 0 else np.nan
        else:
            sv = "N/A"
            speedup_trained_vs_sa = np.nan

        rows.append({
            "hex_id": hex_norm,
            "trained_best": float(tr_best),
            "sa_best": float(sa_best),
            "quality_result": qv,
            "trained_time_s": tr_time,
            "sa_time_s": sa_time,
            "speed_result": sv,
            "speedup_trained_vs_SA": speedup_trained_vs_sa,  # >1  trained faster
            "trained_metrics_csv": str(tr_csv) if tr_csv else "",
            "sa_metrics_csv": str(sa_csv) if sa_csv else "",
            "trained_log": str(tr_log) if tr_log else "",
            "sa_log": str(sa_log) if sa_log else "",
        })

    df = pd.DataFrame(rows)
    n_used = len(df)

    qual_pct = {
        "Trained better %": 100.0 * (df["quality_result"] == "Trained better").sum() / n_used if n_used else 0.0,
        "SA better %":      100.0 * (df["quality_result"] == "SA better").sum()      / n_used if n_used else 0.0,
        "Tie %":            100.0 * (df["quality_result"] == "Tie").sum()            / n_used if n_used else 0.0,
    }

    both_times = df.dropna(subset=["trained_time_s", "sa_time_s"])
    speed_den = len(both_times)
    speed_pct = {
        "Trained faster %": 100.0 * (both_times["speed_result"] == "Trained faster").sum() / speed_den if speed_den else 0.0,
        "SA faster %":      100.0 * (both_times["speed_result"] == "SA faster").sum()      / speed_den if speed_den else 0.0,
        "Tie %":            100.0 * (both_times["speed_result"] == "Tie").sum()            / speed_den if speed_den else 0.0,
    }

    median_speedup = np.nanmedian(df["speedup_trained_vs_SA"]) if not df.empty else np.nan

    summary = pd.DataFrame([{
        "N_used": n_used,
        "Trained_better": (df["quality_result"] == "Trained better").sum(),
        "SA_better":      (df["quality_result"] == "SA better").sum(),
        "Tie":            (df["quality_result"] == "Tie").sum(),
        **qual_pct,
        "Speed_denominator": speed_den,
        "Trained_faster": (both_times["speed_result"] == "Trained faster").sum(),
        "SA_faster":      (both_times["speed_result"] == "SA faster").sum(),
        "Speed_tie":      (both_times["speed_result"] == "Tie").sum(),
        **{f"Speed {k}": v for k, v in speed_pct.items()},
        "median_speedup_trained_vs_SA": median_speedup,
        "N_missing": len(missing),
        "Missing_hex_ids": ",".join(sorted(missing)),
        "Trained_better_hex_ids": ",".join(sorted(trained_better_hex)),
        "SA_better_hex_ids": ",".join(sorted(sa_better_hex)),
        "Tie_hex_ids": ",".join(sorted(tie_hex)),
    }])

    # ---------- NEW: pretty console summaries ----------
    if print_summaries:
        print(f"[Quality] Trained vs SA  N_used={n_used}")
        print(f"  Trained better: {summary.at[0, 'Trained_better']}  "
              f"SA better: {summary.at[0, 'SA_better']}  "
              f"Ties: {summary.at[0, 'Tie']}")
        print(f"  Percentages  Trained better: {qual_pct['Trained better %']:.1f}% | "
              f"SA better: {qual_pct['SA better %']:.1f}% | Tie: {qual_pct['Tie %']:.1f}%")
        if trained_better_hex:
            print(f"  Trained better ({len(trained_better_hex)}): {', '.join(sorted(trained_better_hex))}")
        if sa_better_hex:
            print(f"  SA better ({len(sa_better_hex)}): {', '.join(sorted(sa_better_hex))}")
        if tie_hex:
            print(f"  Ties ({len(tie_hex)}): {', '.join(sorted(tie_hex))}")
        if missing:
            print(f"  Missing ({len(missing)}): {', '.join(sorted(missing))}")

        if speed_den > 0:
            print(f"[Speed] Using circuits with both wall times  N={speed_den}")
            print(f"  Trained faster: {summary.at[0, 'Trained_faster']}  "
                  f"SA faster: {summary.at[0, 'SA_faster']}  "
                  f"Ties: {summary.at[0, 'Speed_tie']}")
            print(f"  Percentages  Trained faster: {speed_pct['Trained faster %']:.1f}% | "
                  f"SA faster: {speed_pct['SA faster %']:.1f}% | Tie: {speed_pct['Tie %']:.1f}%")
            if np.isfinite(median_speedup):
                print(f"  Median speedup (SA_time / trained_time): {median_speedup:.3g}  (>1  trained faster)")
        else:
            print("[Speed] No circuits had both wall times; skipping speed summary.")

    # ---------- plots (unchanged below) ----------
    figs: Dict[str, plt.Figure] = {}

    if make_plots:
        # Quality bar (110  90 pt)
        fig_qbar, ax = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
        cats = ["Trained better %", "SA better %", "Tie %"]
        vals = [qual_pct[c] for c in cats]
        ax.bar(cats, vals, color=["#1976D2", "#8D6E63", "#9E9E9E"], edgecolor=EDGE_COL, linewidth=EDGE_W)
        ax.set_ylim(0, 100)
        ax.set_ylabel("Percentage of circuits")
        ax.set_title(f"Quality: Trained (registry) vs SA (N={n_used})")
        for i, v in enumerate(vals):
            ax.text(i, min(100, v) + 0.5, f"{v:.1f}%", ha="center", va="bottom")
        _style_axes(ax, grid_y=True)
        fig_qbar.tight_layout()
        figs["quality_bar"] = fig_qbar
        if save_quality_bar_path:
            Path(save_quality_bar_path).parent.mkdir(parents=True, exist_ok=True)
            fig_qbar.savefig(save_quality_bar_path, bbox_inches="tight")
        if show_plots:
            plt.show()

        # Quality scatter (100  87.5 pt)
        if not df.empty:
            fig_qs, ax2 = plt.subplots(figsize=(pt_to_in(100), pt_to_in(87.5)), dpi=300)
            ax2.scatter(df["trained_best"], df["sa_best"], s=8, edgecolors=EDGE_COL, linewidths=EDGE_W/2)
            lo = float(min(df["sa_best"].min(), df["trained_best"].min()))
            hi = float(max(df["sa_best"].max(), df["trained_best"].max()))
            ax2.plot([lo, hi], [lo, hi], "--", lw=EDGE_W, color=EDGE_COL)
            ax2.set_ylabel("SA best (lower is better)")
            ax2.set_xlabel("Trained best (lower is better)")
            ax2.set_title("Per-circuit quality")
            ax2.set_aspect("equal", adjustable="box")
            _style_axes(ax2, grid_y=True)
            fig_qs.tight_layout()
            figs["quality_scatter"] = fig_qs
            if save_quality_scatter_path:
                Path(save_quality_scatter_path).parent.mkdir(parents=True, exist_ok=True)
                fig_qs.savefig(save_quality_scatter_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # Speed bar (110  90 pt)
        if speed_den > 0:
            fig_sbar, ax3 = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
            cats_s = ["Trained faster %", "SA faster %", "Tie %"]
            vals_s = [speed_pct[c] for c in cats_s]
            ax3.bar(cats_s, vals_s, color=["#1B5E20", "#C62828", "#9E9E9E"], edgecolor=EDGE_COL, linewidth=EDGE_W)
            ax3.set_ylim(0, 100)
            ax3.set_ylabel(f"Percentage of circuits (N={speed_den})")
            ax3.set_title("Speed comparison")
            for i, v in enumerate(vals_s):
                ax3.text(i, min(100, v) + 0.5, f"{v:.1f}%", ha="center", va="bottom")
            _style_axes(ax3, grid_y=True)
            fig_sbar.tight_layout()
            figs["speed_bar"] = fig_sbar
            if save_speed_bar_path:
                Path(save_speed_bar_path).parent.mkdir(parents=True, exist_ok=True)
                fig_sbar.savefig(save_speed_bar_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # Time scatter (log-log)
        df_times = df.dropna(subset=["trained_time_s", "sa_time_s"])
        if not df_times.empty:
            fig_tsc, ax4 = plt.subplots(figsize=(pt_to_in(100), pt_to_in(87.5)), dpi=300)
            x = df_times["trained_time_s"].astype(float).to_numpy()
            y = df_times["sa_time_s"].astype(float).to_numpy()
            ax4.scatter(x, y, s=8, edgecolors=EDGE_COL, linewidths=EDGE_W/2)
            lo = float(min(np.min(x), np.min(y))); hi = float(max(np.max(x), np.max(y)))
            lo = max(lo, 1e-3)
            ax4.plot([lo, hi], [lo, hi], "--", lw=EDGE_W, color=EDGE_COL)
            ax4.set_xscale("log"); ax4.set_yscale("log")
            ax4.set_ylabel("SA wall time (s)")
            ax4.set_xlabel("Trained wall time (s)")
            ax4.set_title("Per-circuit time (loglog)")
            _style_axes(ax4, grid_y=True)
            fig_tsc.tight_layout()
            figs["time_scatter"] = fig_tsc
            if save_time_scatter_path:
                Path(save_time_scatter_path).parent.mkdir(parents=True, exist_ok=True)
                fig_tsc.savefig(save_time_scatter_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # Per-circuit speedup
        df_speed = df_times.copy()
        if not df_speed.empty:
            df_speed["speedup_trained_vs_SA"] = (df_speed["sa_time_s"] / df_speed["trained_time_s"]).astype(float)
            df_speed = df_speed.dropna(subset=["speedup_trained_vs_SA"]).sort_values("speedup_trained_vs_SA", ascending=False)
        if not df_speed.empty:
            fig_sup, ax5 = plt.subplots(figsize=(pt_to_in(max(150, 6 * len(df_speed))), pt_to_in(90)), dpi=300)
            ax5.bar(df_speed["hex_id"], df_speed["speedup_trained_vs_SA"], color="#455A64",
                    edgecolor=EDGE_COL, linewidth=EDGE_W)
            ax5.axhline(1.0, color=EDGE_COL, ls="--", lw=EDGE_W)
            ax5.set_ylabel("Speedup (trained vs SA) = SA_time / trained_time")
            ax5.set_xlabel("Circuit (hex)")
            ax5.set_title(">1  trained faster")
            ax5.set_xticklabels(df_speed["hex_id"], rotation=60, ha="right")
            _style_axes(ax5, grid_y=True)
            fig_sup.tight_layout()
            figs["speedup_plot"] = fig_sup
            if save_speedup_plot_path:
                Path(save_speedup_plot_path).parent.mkdir(parents=True, exist_ok=True)
                fig_sup.savefig(save_speedup_plot_path, bbox_inches="tight")
            if show_plots:
                plt.show()

    # saves
    if save_summary_csv:
        save_summary_csv = Path(save_summary_csv)
        save_summary_csv.parent.mkdir(parents=True, exist_ok=True)
        summary.to_csv(save_summary_csv, index=False)
        df.to_csv(save_summary_csv.with_name(save_summary_csv.stem + "_per_hex.csv"), index=False)

    return summary, df, figs

```

```python
import re
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib as mpl
from typing import List, Optional, Tuple, Dict
from matplotlib.patches import Patch

# ---------- unified export-friendly style (5 pt) ----------
mpl.rcParams.update({
    "font.size": 5, "axes.titlesize": 5, "axes.labelsize": 5,
    "legend.fontsize": 5, "xtick.labelsize": 5, "ytick.labelsize": 5,
    "pdf.fonttype": 42, "ps.fonttype": 42, "pdf.use14corefonts": False, "svg.fonttype": "none",
    "axes.linewidth": 0.5, "lines.linewidth": 0.5,
    "legend.frameon": False, "savefig.dpi": 300, "figure.dpi": 100,
    "font.family": "DejaVu Sans",
})
EDGE_COL, EDGE_W = "#58595B", 0.5
pt_to_in = lambda pt: pt / 72.0

def _style_axes(ax, grid_y=True):
    ax.tick_params(axis="y", which="major", width=EDGE_W, length=6, color=EDGE_COL)
    ax.tick_params(axis="y", which="minor", width=EDGE_W, length=3, color=EDGE_COL)
    ax.tick_params(axis="x", which="both",  width=EDGE_W, length=3, color=EDGE_COL)
    for s in ["left", "bottom"]:
        ax.spines[s].set_linewidth(EDGE_W)
        ax.spines[s].set_color(EDGE_COL)
    for s in ["top", "right"]:
        ax.spines[s].set_visible(False)
    if grid_y:
        ax.grid(True, axis="y", which="major", linewidth=0.3, alpha=0.3)

# ---------- path templates (edit TRAINED if your layout differs) ----------
# sampling factor 4
TRAINED_ROOT_TEMPLATE = ( 
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/zero-shot/mp/11400_steps/4_inputs/initial_state_sampling_factor_4/{HEX}/seed_1"
)

# no fw preference (SA)
SA_ROOT_TEMPLATE = (
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/logic_synthesis_SA/11400_steps/4_inputs/fw_preference/{HEX}/seed_1"
)

LOG_GLOBS = ["**/slurm*.log", "slurm*.log"]

# ---------- utilities ----------
def _norm_hex(x: str) -> str:
    s = str(x).strip().lower()
    if not s.startswith("0x"): s = "0x" + s
    return "0x" + s[2:].upper()

def trained_root_for_hex(hex_id: str) -> Path:
    return Path(TRAINED_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

def sa_root_for_hex(hex_id: str) -> Path:
    return Path(SA_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

# ---------- trained metrics discovery ----------
TRAINED_CSV_PATTERNS = [
    "trained_masked/*trained_episode_metrics.csv",
    "trained_masked/trajectories/*episode_metrics.csv",
    "**/*trained_episode_metrics.csv",
    "**/trajectories/*episode_metrics.csv",
]

def find_trained_csv(root: Path) -> Optional[Path]:
    for pat in TRAINED_CSV_PATTERNS:
        hits = sorted(set(p for p in root.glob(pat)))
        if hits:
            return hits[0]
    for name in ("trained_episode_metrics.csv",):
        p = root / name
        if p.exists(): return p
    return None

def trained_final_best(csv_path: Path) -> Optional[float]:
    try:
        df = pd.read_csv(csv_path)
        if "episode" in df.columns:
            df = df.sort_values("episode", kind="mergesort")
        if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
            s = pd.to_numeric(df["best_energy_so_far"], errors="coerce").dropna()
            if len(s): return float(s.iloc[-1])
        if "best_energy_in_episode" in df.columns and df["best_energy_in_episode"].notna().any():
            s = pd.to_numeric(df["best_energy_in_episode"], errors="coerce")
            cm = s.cummin().dropna()
            if len(cm): return float(cm.iloc[-1])
    except Exception as e:
        print(f"[WARN] trained read failed {csv_path}: {e}")
    return None

def load_trained_series(csv_path: Optional[Path]) -> Tuple[np.ndarray, np.ndarray]:
    """
    Returns (trained_steps, trained_best_so_far) as numpy arrays.
    Uses 'best_energy_so_far' if present; else cummin('best_energy_in_episode').
    Steps are the 'episode' index (ints).
    """
    if csv_path is None:
        return np.array([], dtype=int), np.array([], dtype=float)
    try:
        df = pd.read_csv(csv_path)
    except Exception as e:
        print(f"[WARN] trained CSV read failed {csv_path}: {e}")
        return np.array([], dtype=int), np.array([], dtype=float)

    if "episode" not in df.columns:
        return np.array([], dtype=int), np.array([], dtype=float)

    df["episode"] = pd.to_numeric(df["episode"], errors="coerce")
    df = df.dropna(subset=["episode"]).sort_values("episode", kind="mergesort")
    steps = df["episode"].astype(int).to_numpy()

    vals: Optional[np.ndarray] = None
    if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
        vals = pd.to_numeric(df["best_energy_so_far"], errors="coerce").to_numpy()
    elif "best_energy_in_episode" in df.columns and df["best_energy_in_episode"].notna().any():
        epi = pd.to_numeric(df["best_energy_in_episode"], errors="coerce").to_numpy()
        vals = np.minimum.accumulate(epi)  # cummin across episodes
    else:
        return np.array([], dtype=int), np.array([], dtype=float)

    vals = np.minimum.accumulate(vals.astype(float))
    return steps, vals

# ---------- SA metrics discovery ----------
def find_sa_csv(root: Path) -> Optional[Path]:
    p = root / "sa_masked" / "sa_step_metrics.csv"
    if p.exists(): return p
    for pat in ["**/sa_step_metrics.csv", "**/*sa_step_metrics.csv"]:
        hits = sorted(set(p for p in root.glob(pat)))
        if hits: return hits[0]
    return None

def sa_final_best(csv_path: Path) -> Optional[float]:
    try:
        df = pd.read_csv(csv_path)
        if "step" in df.columns:
            df["step"] = pd.to_numeric(df["step"], errors="coerce")
            df = df.dropna(subset=["step"]).sort_values("step", kind="mergesort")
        if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
            s = pd.to_numeric(df["best_energy_so_far"], errors="coerce").dropna()
            if len(s): return float(s.iloc[-1])
        if "energy" in df.columns and df["energy"].notna().any():
            e = pd.to_numeric(df["energy"], errors="coerce").dropna()
            if len(e): return float(e.cummin().iloc[-1])
    except Exception as e:
        print(f"[WARN] SA read failed {csv_path}: {e}")
    return None

def load_sa_series(csv_path: Optional[Path]) -> Tuple[np.ndarray, np.ndarray]:
    """
    Returns (sa_steps, sa_best_so_far) as numpy arrays.
    Uses 'best_energy_so_far' if present; else cummin('energy').
    """
    if csv_path is None:
        return np.array([], dtype=int), np.array([], dtype=float)
    try:
        df = pd.read_csv(csv_path)
    except Exception as e:
        print(f"[WARN] SA CSV read failed {csv_path}: {e}")
        return np.array([], dtype=int), np.array([], dtype=float)

    if "step" not in df.columns:
        return np.array([], dtype=int), np.array([], dtype=float)

    df["step"] = pd.to_numeric(df["step"], errors="coerce")
    df = df.dropna(subset=["step"]).sort_values("step", kind="mergesort")
    steps = df["step"].astype(int).to_numpy()

    if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
        vals = pd.to_numeric(df["best_energy_so_far"], errors="coerce").to_numpy()
    elif "energy" in df.columns and df["energy"].notna().any():
        eng = pd.to_numeric(df["energy"], errors="coerce").to_numpy()
        vals = np.minimum.accumulate(eng)
    else:
        return np.array([], dtype=int), np.array([], dtype=float)

    vals = np.minimum.accumulate(vals.astype(float))
    return steps, vals

# ---------- log parsing ----------
_sa_total_re = re.compile(r"Total time\s*=\s*([\d\.]+)s", re.IGNORECASE)
_generic_hms_re = re.compile(r"Completed in\s*(\d+)h\s+(\d+)m\s+(\d+)s", re.IGNORECASE)
_trained_timer_re = re.compile(r"\[TIMER\]\s*Track\s*1\s*run\s*time:\s*([\d\.]+)s", re.IGNORECASE)

def _find_latest_log(root: Path) -> Optional[Path]:
    hits = []
    for pat in LOG_GLOBS:
        hits.extend(root.glob(pat))
    if not hits: return None
    hits = sorted(hits, key=lambda p: p.stat().st_mtime, reverse=True)
    return hits[0]

def parse_sa_time_seconds(log_path: Path) -> Optional[float]:
    try:
        txt = log_path.read_text(errors="ignore")
        m = None
        for m in _sa_total_re.finditer(txt):  # prefer "Total time=...s"
            pass
        if m:
            return float(m.group(1))
        m2 = None
        for m2 in _generic_hms_re.finditer(txt):  # fallback "Completed in Hh Mm Ss"
            pass
        if m2:
            h, m, s = map(int, m2.groups())
            return h*3600 + m*60 + s
    except Exception as e:
        print(f"[WARN] SA time parse failed {log_path}: {e}")
    return None

def parse_trained_time_seconds(log_path: Path) -> Optional[float]:
    """
    Prefer '[TIMER] Track 1 run time: <secs>s' if present; fall back to 'Completed in Hh Mm Ss'
    or 'Total time=<secs>s' (rare but supported).
    """
    try:
        txt = log_path.read_text(errors="ignore")

        # Preferred
        m = None
        for m in _trained_timer_re.finditer(txt):
            pass
        if m:
            return float(m.group(1))

        # Fallback: Completed in 0h 14m 2s
        m2 = None
        for m2 in _generic_hms_re.finditer(txt):
            pass
        if m2:
            h, mm, ss = map(int, m2.groups())
            return h*3600 + mm*60 + ss

        # Last resort: Total time=123.4s
        m3 = None
        for m3 in _sa_total_re.finditer(txt):
            pass
        if m3:
            return float(m3.group(1))

    except Exception as e:
        print(f"[WARN] trained time parse failed {log_path}: {e}")
    return None

def _best_so_far(arr: np.ndarray) -> np.ndarray:
    if arr.size == 0:
        return arr
    return np.minimum.accumulate(arr)

# ---------- main comparer ----------
def compare_trained_vs_sa(
    hex_ids: List[str],
    make_plots: bool = True,
    show_plots: bool = True,
    save_quality_bar_path: Optional[Path] = None,
    save_quality_scatter_path: Optional[Path] = None,
    save_speed_bar_path: Optional[Path] = None,
    save_time_scatter_path: Optional[Path] = None,
    save_speedup_plot_path: Optional[Path] = None,
    save_summary_csv: Optional[Path] = None,
    # Matched-steps figure (SA-equivalent steps; episode10):
    save_pct_over_steps_path: Optional[Path] = None,
    step_bin: int = 1000,
    # Per-circuit runtime plot
    save_runtime_bar_path: Optional[Path] = None,
    # console
    print_summaries: bool = True,
):
    """
    Quality: lower energy is better.
    Speed: lower wall time (s) is faster.
    Speedup (trained vs SA): SA_time / trained_time  (>1  trained faster).

    Matched-steps plot matches on **SA-equivalent steps**:
      trained episode k  10 * k SA steps (episodes scaled by 10 before binning).
    """
    rows = []
    missing = []

    quality_counts = {"Trained better": 0, "SA better": 0, "Tie": 0}
    trained_better_hex, sa_better_hex, tie_hex = [], [], []

    speed_counts = {"Trained faster": 0, "SA faster": 0, "Tie": 0}
    speed_den = 0

    # collect series per circuit for matched-steps plot (store trained as SA-equivalent steps)
    series_by_hex: Dict[str, Tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]] = {}

    for hex_id in hex_ids:
        hex_norm = _norm_hex(hex_id)
        tr_root = trained_root_for_hex(hex_norm)
        sa_root = sa_root_for_hex(hex_norm)

        tr_csv = find_trained_csv(tr_root)
        sa_csv = find_sa_csv(sa_root)
        tr_best = trained_final_best(tr_csv) if tr_csv else None
        sa_best = sa_final_best(sa_csv) if sa_csv else None

        # Load series for matched-steps
        tr_ep, tr_vals = load_trained_series(tr_csv)  # episodes
        sa_steps, sa_vals = load_sa_series(sa_csv)    # SA steps

        if tr_ep.size and sa_steps.size:
            # Convert episodes  SA-equivalent steps (episode k = 10*k)
            tr_steps_eq = (tr_ep.astype(int) * 10).astype(int)

            # Ensure sorted + best-so-far (robustness)
            tr_ord = np.argsort(tr_steps_eq, kind="mergesort")
            sa_ord = np.argsort(sa_steps,    kind="mergesort")
            tr_steps_eq = tr_steps_eq[tr_ord]
            tr_vals     = _best_so_far(np.asarray(tr_vals)[tr_ord].astype(float))
            sa_steps    = sa_steps[sa_ord].astype(int)
            sa_vals     = _best_so_far(np.asarray(sa_vals)[sa_ord].astype(float))

            series_by_hex[hex_norm] = (tr_steps_eq, tr_vals, sa_steps, sa_vals)

        tr_log = _find_latest_log(tr_root)
        sa_log = _find_latest_log(sa_root)
        tr_time = parse_trained_time_seconds(tr_log) if tr_log else None
        sa_time = parse_sa_time_seconds(sa_log) if sa_log else None

        if tr_best is None or sa_best is None:
            missing.append(hex_norm)
            continue

        # Quality verdict
        if tr_best < sa_best:
            qv = "Trained better"; quality_counts["Trained better"] += 1; trained_better_hex.append(hex_norm)
        elif tr_best > sa_best:
            qv = "SA better";      quality_counts["SA better"] += 1;      sa_better_hex.append(hex_norm)
        else:
            qv = "Tie";            quality_counts["Tie"] += 1;            tie_hex.append(hex_norm)

        # Speed verdict & folds
        if (tr_time is not None) and (sa_time is not None):
            speed_den += 1
            if tr_time < sa_time:
                sv = "Trained faster"; speed_counts["Trained faster"] += 1
            elif tr_time > sa_time:
                sv = "SA faster";      speed_counts["SA faster"] += 1
            else:
                sv = "Tie";            speed_counts["Tie"] += 1
            speedup_trained_vs_sa = float(sa_time) / float(tr_time) if tr_time and tr_time > 0 else np.nan
        else:
            sv = "N/A"
            speedup_trained_vs_sa = np.nan

        rows.append({
            "hex_id": hex_norm,
            "trained_best": float(tr_best),
            "sa_best": float(sa_best),
            "quality_result": qv,
            "trained_time_s": tr_time,
            "sa_time_s": sa_time,
            "speed_result": sv,
            "speedup_trained_vs_SA": speedup_trained_vs_sa,  # >1  trained faster
            "trained_metrics_csv": str(tr_csv) if tr_csv else "",
            "sa_metrics_csv": str(sa_csv) if sa_csv else "",
            "trained_log": str(tr_log) if tr_log else "",
            "sa_log": str(sa_log) if sa_log else "",
            # stash (trained stored as SA-equivalent steps)
            "_trained_steps_eq": series_by_hex.get(hex_norm, (np.array([], int),))[0] if hex_norm in series_by_hex else np.array([], int),
            "_trained_vals":     series_by_hex.get(hex_norm, (None, np.array([], float)))[1] if hex_norm in series_by_hex else np.array([], float),
            "_sa_steps":         series_by_hex.get(hex_norm, (None, None, np.array([], int)))[2] if hex_norm in series_by_hex else np.array([], int),
            "_sa_vals":          series_by_hex.get(hex_norm, (None, None, None, np.array([], float)))[3] if hex_norm in series_by_hex else np.array([], float),
        })

    df = pd.DataFrame(rows)
    n_used = len(df)

    qual_pct = {
        "Trained better %": 100.0 * (df["quality_result"] == "Trained better").sum() / n_used if n_used else 0.0,
        "SA better %":      100.0 * (df["quality_result"] == "SA better").sum()      / n_used if n_used else 0.0,
        "Tie %":            100.0 * (df["quality_result"] == "Tie").sum()            / n_used if n_used else 0.0,
    }

    both_times = df.dropna(subset=["trained_time_s", "sa_time_s"])
    speed_den = len(both_times)
    speed_pct = {
        "Trained faster %": 100.0 * (both_times["speed_result"] == "Trained faster").sum() / speed_den if speed_den else 0.0,
        "SA faster %":      100.0 * (both_times["speed_result"] == "SA faster").sum()      / speed_den if speed_den else 0.0,
        "Tie %":            100.0 * (both_times["speed_result"] == "Tie").sum()            / speed_den if speed_den else 0.0,
    }

    median_speedup = np.nanmedian(df["speedup_trained_vs_SA"]) if not df.empty else np.nan

    summary = pd.DataFrame([{
        "N_used": n_used,
        "Trained_better": (df["quality_result"] == "Trained better").sum(),
        "SA_better":      (df["quality_result"] == "SA better").sum(),
        "Tie":            (df["quality_result"] == "Tie").sum(),
        **qual_pct,
        "Speed_denominator": speed_den,
        "Trained_faster": (both_times["speed_result"] == "Trained faster").sum(),
        "SA_faster":      (both_times["speed_result"] == "SA faster").sum(),
        "Speed_tie":      (both_times["speed_result"] == "Tie").sum(),
        "median_speedup_trained_vs_SA": median_speedup,
        "N_missing": len(missing),
        "Missing_hex_ids": ",".join(sorted(missing)),
        "Trained_better_hex_ids": ",".join(sorted(trained_better_hex)),
        "SA_better_hex_ids": ",".join(sorted(sa_better_hex)),
        "Tie_hex_ids": ",".join(sorted(tie_hex)),
    }])

    # ---------- Pretty console summaries incl. circuits per bucket ----------
    if print_summaries:
        print(f"[Quality] Trained vs SA  N_used={n_used}")
        print(f"  Trained better: {summary.at[0, 'Trained_better']}  "
              f"SA better: {summary.at[0, 'SA_better']}  "
              f"Ties: {summary.at[0, 'Tie']}")
        print(f"  Percentages  Trained better: {qual_pct['Trained better %']:.1f}% | "
              f"SA better: {qual_pct['SA better %']:.1f}% | Tie: {qual_pct['Tie %']:.1f}%")
        if trained_better_hex:
            print("  Circuits WON by Trained: ", ", ".join(sorted(trained_better_hex)))
        if sa_better_hex:
            print("  Circuits LOST to SA:     ", ", ".join(sorted(sa_better_hex)))
        if tie_hex:
            print("  Circuits TIED:           ", ", ".join(sorted(tie_hex)))
        if missing:
            print("  Circuits MISSING:        ", ", ".join(sorted(missing)))

        if speed_den > 0:
            print(f"[Speed] Using circuits with both wall times  N={speed_den}")
            print(f"  Trained faster: {(both_times['speed_result'] == 'Trained faster').sum()}  "
                  f"SA faster: {(both_times['speed_result'] == 'SA faster').sum()}  "
                  f"Ties: {(both_times['speed_result'] == 'Tie').sum()}")
            print(f"  Percentages  Trained faster: {speed_pct['Trained faster %']:.1f}% | "
                  f"SA faster: {speed_pct['SA faster %']:.1f}% | Tie: {speed_pct['Tie %']:.1f}%")
            if np.isfinite(median_speedup):
                print(f"  Median speedup (SA_time / trained_time): {median_speedup:.3g}  (>1  trained faster)")
        else:
            print("[Speed] No circuits had both wall times; skipping speed summary.")

    # ---------- plots ----------
    figs: Dict[str, plt.Figure] = {}

    if make_plots:
        # 1) Quality bar
        fig_qbar, ax = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
        cats = ["Trained better %", "SA better %", "Tie %"]
        vals = [qual_pct[c] for c in cats]
        ax.bar(cats, vals, color=["#1976D2", "#8D6E63", "#9E9E9E"], edgecolor=EDGE_COL, linewidth=EDGE_W)
        ax.set_ylim(0, 100)
        ax.set_ylabel("Percentage of circuits")
        ax.set_title(f"Quality: Trained (registry) vs SA (N={n_used})")
        for i, v in enumerate(vals):
            ax.text(i, min(100, v) + 0.5, f"{v:.1f}%", ha="center", va="bottom")
        _style_axes(ax, grid_y=True)
        fig_qbar.tight_layout()
        figs["quality_bar"] = fig_qbar
        if save_quality_bar_path:
            Path(save_quality_bar_path).parent.mkdir(parents=True, exist_ok=True)
            fig_qbar.savefig(save_quality_bar_path, bbox_inches="tight")
        if show_plots:
            plt.show()

        # 2) Quality scatter
        if not df.empty:
            fig_qs, ax2 = plt.subplots(figsize=(pt_to_in(100), pt_to_in(87.5)), dpi=300)
            ax2.scatter(df["trained_best"], df["sa_best"], s=8, edgecolors=EDGE_COL, linewidths=EDGE_W/2)
            lo = float(min(df["sa_best"].min(), df["trained_best"].min()))
            hi = float(max(df["sa_best"].max(), df["trained_best"].max()))
            ax2.plot([lo, hi], [lo, hi], "--", lw=EDGE_W, color=EDGE_COL)
            ax2.set_ylabel("SA best (lower is better)")
            ax2.set_xlabel("Trained best (lower is better)")
            ax2.set_title("Per-circuit quality")
            ax2.set_aspect("equal", adjustable="box")
            _style_axes(ax2, grid_y=True)
            fig_qs.tight_layout()
            figs["quality_scatter"] = fig_qs
            if save_quality_scatter_path:
                Path(save_quality_scatter_path).parent.mkdir(parents=True, exist_ok=True)
                fig_qs.savefig(save_quality_scatter_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # 3) Speed bar (counts/percent over circuits with both times)
        if speed_den > 0:
            fig_sbar, ax3 = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
            cats_s = ["Trained faster %", "SA faster %", "Tie %"]
            vals_s = [speed_pct[c] for c in cats_s]
            ax3.bar(cats_s, vals_s, color=["#1B5E20", "#C62828", "#9E9E9E"], edgecolor=EDGE_COL, linewidth=EDGE_W)
            ax3.set_ylim(0, 100)
            ax3.set_ylabel(f"Percentage of circuits (N={speed_den})")
            ax3.set_title("Speed comparison")
            for i, v in enumerate(vals_s):
                ax3.text(i, min(100, v) + 0.5, f"{v:.1f}%", ha="center", va="bottom")
            _style_axes(ax3, grid_y=True)
            fig_sbar.tight_layout()
            figs["speed_bar"] = fig_sbar
            if save_speed_bar_path:
                Path(save_speed_bar_path).parent.mkdir(parents=True, exist_ok=True)
                fig_sbar.savefig(save_speed_bar_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # 4) Time scatter (loglog)
        df_times = df.dropna(subset=["trained_time_s", "sa_time_s"])
        if not df_times.empty:
            fig_tsc, ax4 = plt.subplots(figsize=(pt_to_in(100), pt_to_in(87.5)), dpi=300)
            x = df_times["trained_time_s"].astype(float).to_numpy()
            y = df_times["sa_time_s"].astype(float).to_numpy()
            ax4.scatter(x, y, s=8, edgecolors=EDGE_COL, linewidths=EDGE_W/2)
            lo = float(min(np.min(x), np.min(y))); hi = float(max(np.max(x), np.max(y)))
            lo = max(lo, 1e-3)
            ax4.plot([lo, hi], [lo, hi], "--", lw=EDGE_W, color=EDGE_COL)
            ax4.set_xscale("log"); ax4.set_yscale("log")
            ax4.set_ylabel("SA wall time (s)")
            ax4.set_xlabel("Trained wall time (s)")
            ax4.set_title("Per-circuit time (loglog)")
            _style_axes(ax4, grid_y=True)
            fig_tsc.tight_layout()
            figs["time_scatter"] = fig_tsc
            if save_time_scatter_path:
                Path(save_time_scatter_path).parent.mkdir(parents=True, exist_ok=True)
                fig_tsc.savefig(save_time_scatter_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # 5) Per-circuit speedup (SA_time / trained_time)
        df_speed = df_times.copy()
        if not df_speed.empty:
            df_speed["speedup_trained_vs_SA"] = (df_speed["sa_time_s"] / df_speed["trained_time_s"]).astype(float)
            df_speed = df_speed.dropna(subset=["speedup_trained_vs_SA"]).sort_values("speedup_trained_vs_SA", ascending=False)
        if not df_speed.empty:
            fig_sup, ax5 = plt.subplots(figsize=(pt_to_in(max(150, 6 * len(df_speed))), pt_to_in(90)), dpi=300)
            ax5.bar(df_speed["hex_id"], df_speed["speedup_trained_vs_SA"], color="#455A64",
                    edgecolor=EDGE_COL, linewidth=EDGE_W)
            ax5.axhline(1.0, color=EDGE_COL, ls="--", lw=EDGE_W)
            ax5.set_ylabel("Speedup (trained vs SA) = SA_time / trained_time")
            ax5.set_xlabel("Circuit (hex)")
            ax5.set_title(">1  trained faster")
            ax5.set_xticklabels(df_speed["hex_id"], rotation=60, ha="right")
            _style_axes(ax5, grid_y=True)
            fig_sup.tight_layout()
            figs["speedup_plot"] = fig_sup
            if save_speedup_plot_path:
                Path(save_speedup_plot_path).parent.mkdir(parents=True, exist_ok=True)
                fig_sup.savefig(save_speedup_plot_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # 6) Per-circuit runtime grouped bar (Trained vs SA, seconds)
        if not df_times.empty:
            df_rt = df_times.copy()
            # Sort by the slower of the two (descending) so long bars on left
            sort_key = np.maximum(df_rt["trained_time_s"].astype(float), df_rt["sa_time_s"].astype(float))
            df_rt = df_rt.iloc[np.argsort(-sort_key.to_numpy())]
            n = len(df_rt)
            x = np.arange(n, dtype=float)

            fig_rt, axr = plt.subplots(figsize=(pt_to_in(max(180, 10 * n)), pt_to_in(100)), dpi=300)
            bw = 0.42
            axr.bar(x - bw/2, df_rt["trained_time_s"].astype(float).to_numpy(), width=bw,
                    label="Trained", color="#1976D2", edgecolor=EDGE_COL, linewidth=EDGE_W)
            axr.bar(x + bw/2, df_rt["sa_time_s"].astype(float).to_numpy(), width=bw,
                    label="SA", color="#E65100", edgecolor=EDGE_COL, linewidth=EDGE_W)

            axr.set_xticks(x)
            axr.set_xticklabels(df_rt["hex_id"], rotation=60, ha="right")
            axr.set_ylabel("Wall time (s)")
            axr.set_title(f"Per-circuit runtime (lower is faster)  N={n}")
            axr.legend(frameon=False)
            _style_axes(axr, grid_y=True)
            fig_rt.tight_layout()
            figs["runtime_per_circuit_bar"] = fig_rt
            if save_runtime_bar_path:
                Path(save_runtime_bar_path).parent.mkdir(parents=True, exist_ok=True)
                fig_rt.savefig(save_runtime_bar_path, bbox_inches="tight")
            if show_plots:
                plt.show()

        # 7) Matched-steps (SA-equivalent)  episodes10
        if series_by_hex:
            global_max_bin = 0
            for hex_norm, (tr_eq, _, sa_s, _) in series_by_hex.items():
                if tr_eq.size == 0 or sa_s.size == 0:
                    continue
                candidate = min(int(tr_eq[-1]), int(sa_s[-1]))
                if candidate > global_max_bin:
                    global_max_bin = candidate

            if global_max_bin >= step_bin:
                bins = np.arange(step_bin, global_max_bin + step_bin, step_bin, dtype=int)

                tr_better_pct, sa_better_pct, tie_pct, den = [], [], [], []
                for B in bins:
                    tr_better = sa_better = tie = denom = 0
                    for _, (tr_eq, tr_v, sa_s, sa_v) in series_by_hex.items():
                        if tr_eq.size == 0 or sa_s.size == 0:
                            continue
                        i_tr = np.searchsorted(tr_eq, B, side="right") - 1
                        i_sa = np.searchsorted(sa_s,  B, side="right") - 1
                        if i_tr < 0 or i_sa < 0:
                            continue
                        denom += 1
                        tr_val = float(tr_v[i_tr])
                        sa_val = float(sa_v[i_sa])
                        if tr_val < sa_val:
                            tr_better += 1
                        elif tr_val > sa_val:
                            sa_better += 1
                        else:
                            tie += 1
                    den.append(denom)
                    if denom > 0:
                        tr_better_pct.append(100.0 * tr_better / denom)
                        sa_better_pct.append(100.0 * sa_better / denom)
                        tie_pct.append(100.0 * tie / denom)
                    else:
                        tr_better_pct.append(0.0)
                        sa_better_pct.append(0.0)
                        tie_pct.append(0.0)

                fig_pct, ax6 = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
                ax6.plot(bins, tr_better_pct, label="Trained better %")
                ax6.plot(bins, sa_better_pct, label="SA better %")
                ax6.plot(bins, tie_pct,       label="Tie %")
                ax6.set_ylim(0, 100)
                ax6.set_xlabel(f"SA-equivalent steps (episode10), bins of {step_bin}")
                ax6.set_ylabel("Percentage of circuits")
                last_den = den[-1] if len(den) else 0
                ax6.set_title(f"Matched at SA-equivalent steps (N up to {last_den} per bin)")
                _style_axes(ax6, grid_y=True)
                ax6.legend(frameon=False)
                fig_pct.tight_layout()
                figs["percentages_over_matched_steps"] = fig_pct
                if save_pct_over_steps_path:
                    Path(save_pct_over_steps_path).parent.mkdir(parents=True, exist_ok=True)
                    fig_pct.savefig(save_pct_over_steps_path, bbox_inches="tight")
                if show_plots:
                    plt.show()

    # Save CSVs
    if save_summary_csv:
        save_summary_csv = Path(save_summary_csv)
        save_summary_csv.parent.mkdir(parents=True, exist_ok=True)
        summary.to_csv(save_summary_csv, index=False)
        # drop big arrays before saving per-hex
        drop_cols = [c for c in ["_trained_steps_eq","_trained_vals","_sa_steps","_sa_vals"] if c in df.columns]
        df_to_save = df.drop(columns=drop_cols) if drop_cols else df
        df_to_save.to_csv(save_summary_csv.with_name(save_summary_csv.stem + "_per_hex.csv"), index=False)

    return summary, df, figs


```

```python
# ---------- TRAINED (CSV) vs FINE-TUNING (TensorBoard) ----------
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib as mpl
from typing import Dict, List, Tuple, Optional
from tensorboard.backend.event_processing import event_accumulator

# ====== you already have these in scope ======
# hex_ids: List[str]

# ---------- unified, export-friendly style (5 pt) ----------
mpl.rcParams.update({
    "font.family": "DejaVu Sans",
    "font.size": 5, "axes.titlesize": 5, "axes.labelsize": 5,
    "legend.fontsize": 5, "xtick.labelsize": 5, "ytick.labelsize": 5,
    "pdf.fonttype": 42, "ps.fonttype": 42, "pdf.use14corefonts": False, "svg.fonttype": "none",
    "axes.linewidth": 0.5, "lines.linewidth": 0.5,
    "legend.frameon": False, "savefig.dpi": 300, "figure.dpi": 100,
})
EDGE_COL, EDGE_W = "#58595B", 0.5
pt_to_in = lambda pt: pt / 72.0

def _style_axes(ax, grid_y=True):
    ax.tick_params(axis="y", which="major", width=EDGE_W, length=6, color=EDGE_COL)
    ax.tick_params(axis="y", which="minor", width=EDGE_W, length=3, color=EDGE_COL)
    ax.tick_params(axis="x", which="both",  width=EDGE_W, length=3, color=EDGE_COL)
    for s in ["left", "bottom"]:
        ax.spines[s].set_linewidth(EDGE_W)
        ax.spines[s].set_color(EDGE_COL)
    for s in ["top", "right"]:
        ax.spines[s].set_visible(False)
    if grid_y:
        ax.grid(True, axis="y", which="major", linewidth=0.3, alpha=0.3)

# ---- TB config for FINE-TUNING ----
FINETUNE_ROOT_TEMPLATE = (
    
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/fine_tune/GAT_MLP_with_scalars/4000nn/initial_state_sampling_factor_0/4_inputs/50_000_steps/{HEX}/seed_1"
)
EVENT_GLOBS = ["**/events.out.tfevents.*", "**/*tfevents*"]
TB_TAG = "custom/best_energy"
ASSUME_TAG_IS_BEST = True  # True: compare final point; False: compare global min over series

# ---- Normalization ----
def _norm_hex(x: str) -> str:
    s = str(x).strip().lower()
    if not s.startswith("0x"): s = "0x" + s
    return "0x" + s[2:].upper()

# ---- TRAINED CSV discovery patterns ----
TRAINED_CSV_PATTERNS = [
    "trained_masked/*trained_episode_metrics.csv",
    "trained_masked/trajectories/*episode_metrics.csv",
    "**/*trained_episode_metrics.csv",
    "**/trajectories/*episode_metrics.csv",
    "trained_episode_metrics.csv",  # fallback at root
]

def _trained_root_for_hex(hex_id: str) -> Path:
    # Adjust to your actual trained directory layout if needed
    return Path(
        "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
        f"manuscript/zero-shot/mp/5120_episodes/sf4/{_norm_hex(hex_id)}/seed_1"
    )

def _finetune_root_for_hex(hex_id: str) -> Path:
    return Path(FINETUNE_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

def find_trained_csv(search_root: Path) -> Optional[Path]:
    root = Path(search_root)
    for pat in TRAINED_CSV_PATTERNS:
        hits = sorted(set(p for p in root.glob(pat)))
        if hits:
            return hits[0]
    return None

def get_trained_final_best(csv_path: Path) -> Optional[float]:
    """
    For trained per-episode metrics:
      - prefers 'best_energy_so_far'
      - else uses cummin('best_energy_in_episode')
      - returns the final best value (lower is better)
    """
    try:
        df = pd.read_csv(csv_path)
        if "episode" in df.columns:
            df["episode"] = pd.to_numeric(df["episode"], errors="coerce")
            df = df.dropna(subset=["episode"]).sort_values("episode", kind="mergesort")
        if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
            s = pd.to_numeric(df["best_energy_so_far"], errors="coerce").dropna()
            if len(s) > 0: return float(s.iloc[-1])
        if "best_energy_in_episode" in df.columns and df["best_energy_in_episode"].notna().any():
            e = pd.to_numeric(df["best_energy_in_episode"], errors="coerce").dropna()
            if len(e) > 0: return float(e.cummin().iloc[-1])
    except Exception as e:
        print(f"[WARN] Failed to read trained CSV {csv_path}: {e}")
    return None

# ---- Fine-tuning TensorBoard helpers ----
def find_event_files(root: Path) -> List[Path]:
    files = []
    for pat in EVENT_GLOBS: files.extend(root.glob(pat))
    return sorted(set(files), key=lambda p: p.as_posix())

def load_tb_scalar_from_dir(run_dir: Path, tag: str) -> Tuple[np.ndarray, np.ndarray]:
    try:
        ea = event_accumulator.EventAccumulator(str(run_dir)); ea.Reload()
        if tag not in ea.Tags().get("scalars", []):
            return np.array([], dtype=int), np.array([], dtype=float)
        ev = ea.Scalars(tag)
        steps  = np.fromiter((e.step  for e in ev), dtype=int)
        values = np.fromiter((e.value for e in ev), dtype=float)
        return steps, values
    except Exception as e:
        print(f"[WARN] TB load failed in {run_dir}: {e}")
        return np.array([], dtype=int), np.array([], dtype=float)

def pick_best_from_runs(run_dirs: List[Path], tag: str, assume_best_series=True) -> Tuple[Optional[float], Optional[Path], np.ndarray, np.ndarray]:
    """Return (best_value, chosen_dir, steps, values) for the best run (lowest value)."""
    best_val, best_dir, best_steps, best_values = None, None, None, None
    for d in run_dirs:
        steps, vals = load_tb_scalar_from_dir(d, tag)
        if steps.size == 0: continue
        candidate = float(vals[-1]) if assume_best_series else float(np.nanmin(vals))
        if best_val is None or candidate < best_val:
            best_val, best_dir, best_steps, best_values = candidate, d, steps, vals
    return best_val, best_dir, best_steps, best_values

# ======== Plot helper (compact bar) ========
def _bar_figure_compact(categories: List[str], values: List[float], title: str) -> plt.Figure:
    fig, ax = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
    bars = ax.bar(categories, values, color=["#1976D2", "#8D6E63", "#9E9E9E"],
                  edgecolor=EDGE_COL, linewidth=EDGE_W)
    ax.set_ylim(0, 100)
    ax.set_ylabel("Percentage of circuits")
    ax.set_title(title)
    for i, v in enumerate(values):
        ax.text(i, min(100, v) + 0.5, f"{v:.1f}%", ha="center", va="bottom")
    _style_axes(ax, grid_y=True)
    fig.tight_layout()
    return fig

def _to_percentages(counts: Dict[str, int], denom: int) -> List[float]:
    if denom == 0: return [0.0, 0.0, 0.0]
    return [100.0*counts["Win"]/denom, 100.0*counts["Loss"]/denom, 100.0*counts["Tie"]/denom]

# ======== Core comparison: TRAINED vs FINE-TUNING ========
def compare_trained_to_finetune(
    hex_ids: List[str],
    tb_tag: str = TB_TAG,
    assume_tag_is_best: bool = ASSUME_TAG_IS_BEST,
    return_rows: bool = True,
):
    """
    Compare the final best energies between:
      - TRAINED: per-episode CSV (best_energy_so_far or cummin(best_energy_in_episode))
      - FINETUNE: TensorBoard scalar series (tag)
    Lower is better. 'Win' means TRAINED < FINETUNE.
    """
    outcomes = {"Win": 0, "Loss": 0, "Tie": 0}  # Win = TRAINED better
    missing, rows = [], []
    trained_hex, finetune_hex, tie_hex = [], [], []

    for hex_id in hex_ids:
        hex_norm = _norm_hex(hex_id)

        # TRAINED (CSV)
        tr_root = _trained_root_for_hex(hex_norm)
        tr_csv = find_trained_csv(tr_root)
        tr_final = get_trained_final_best(tr_csv) if tr_csv else None

        # FINETUNE (TensorBoard)
        ft_root = _finetune_root_for_hex(hex_norm)
        ft_events = find_event_files(ft_root)
        ft_run_dirs = sorted(set(p.parent for p in ft_events))
        ft_best, ft_dir, ft_steps, ft_vals = pick_best_from_runs(ft_run_dirs, tb_tag, assume_tag_is_best)

        if tr_final is None or ft_best is None:
            missing.append(hex_norm)
            continue

        if tr_final < ft_best:
            outcomes["Win"] += 1; verdict = "Win";  trained_hex.append(hex_norm)
        elif tr_final > ft_best:
            outcomes["Loss"] += 1; verdict = "Loss"; finetune_hex.append(hex_norm)
        else:
            outcomes["Tie"] += 1; verdict = "Tie";  tie_hex.append(hex_norm)

        if return_rows:
            rows.append({
                "hex_id": hex_norm,
                "trained_final": float(tr_final),
                "finetune_best": float(ft_best),
                "delta_trained_minus_finetune": float(tr_final - ft_best),
                "result": verdict,
                "trained_csv": str(tr_csv) if tr_csv else "",
                "finetune_chosen_run_dir": str(ft_dir) if ft_dir else "",
            })

    n_used = len(rows)
    if n_used == 0:
        win_pct = loss_pct = tie_pct = 0.0
    else:
        win_pct, loss_pct, tie_pct = _to_percentages(outcomes, n_used)

    # Pretty console summary
    print(f"[Trained vs Fine-tuning] counts: {outcomes} out of N={n_used} (missing {len(missing)})")
    if trained_hex:
        print(f"  Trained better ({len(trained_hex)}): {', '.join(sorted(trained_hex))}")
    if finetune_hex:
        print(f"  Fine-tuning better ({len(finetune_hex)}): {', '.join(sorted(finetune_hex))}")
    if tie_hex:
        print(f"  Ties ({len(tie_hex)}): {', '.join(sorted(tie_hex))}")
    if missing:
        print(f"  Missing ({len(missing)}): {', '.join(sorted(missing))}")

    per_hex_df = pd.DataFrame(rows)
    summary_df = pd.DataFrame([{
        "method": "trained_vs_finetune",
        "label": "Trained policy vs Fine-tuning",
        "N_used": n_used,
        "Win": outcomes["Win"], "Loss": outcomes["Loss"], "Tie": outcomes["Tie"],
        "Win_%": win_pct, "Loss_%": loss_pct, "Tie_%": tie_pct,
        "N_missing": len(missing),
        "Missing_hex_ids": ",".join(sorted(missing)),
        "Trained_better_hex_ids": ",".join(sorted(trained_hex)),
        "Finetune_better_hex_ids": ",".join(sorted(finetune_hex)),
        "Tie_hex_ids": ",".join(sorted(tie_hex)),
        "tb_tag": tb_tag,
        "assume_tag_is_best": assume_tag_is_best,
    }])

    return summary_df, per_hex_df

# ======== Orchestrator (TRAINED vs FINE-TUNING) ========
def summarize_and_plot_trained_vs_finetune(
    hex_ids: List[str],
    tb_tag: str = TB_TAG,
    assume_tag_is_best: bool = ASSUME_TAG_IS_BEST,
    make_plots: bool = True,
    show_plots: bool = False,
    save_summary_csv: Optional[Path] = None,
    save_plot_path: Optional[Path] = None,
):
    """
    Returns (summary_df, per_hex_df, fig_or_None)
    """
    cats = ["Win", "Loss", "Tie"]
    summary, per_hex = compare_trained_to_finetune(hex_ids, tb_tag, assume_tag_is_best, return_rows=True)
    win_pct, loss_pct, tie_pct = float(summary["Win_%"].iloc[0]), float(summary["Loss_%"].iloc[0]), float(summary["Tie_%"].iloc[0])

    fig = None
    if make_plots:
        title = f"Trained vs Fine-tuning (N={int(summary['N_used'].iloc[0])})  tag: {tb_tag}"
        fig = _bar_figure_compact(cats, [win_pct, loss_pct, tie_pct], title)
        if save_plot_path is not None:
            Path(save_plot_path).parent.mkdir(parents=True, exist_ok=True)
            fig.savefig(save_plot_path, bbox_inches="tight")
        if show_plots:
            plt.show()

    if save_summary_csv is not None:
        Path(save_summary_csv).parent.mkdir(parents=True, exist_ok=True)
        summary.to_csv(save_summary_csv, index=False)
        per_hex_out = save_summary_csv.with_name(save_summary_csv.stem + "_per_hex.csv")
        per_hex.to_csv(per_hex_out, index=False)

    return summary, per_hex, fig

# ======================== example usage ========================
# summary_df, per_hex_df, fig = summarize_and_plot_trained_vs_finetune(
#     hex_ids,
#     tb_tag="custom/best_energy",   # or your desired TB scalar tag
#     assume_tag_is_best=True,       # True: compare last point; False: compare global min
#     make_plots=True,
#     show_plots=True,
#     # save_summary_csv=Path("./trained_vs_finetune_summary.csv"),
#     # save_plot_path=Path("./trained_vs_finetune.png"),
# )


# ======================== example usage ========================
summary_df, per_hex_df, fig = summarize_and_plot_trained_vs_finetune(
    hex_ids,
    tb_tag="custom/best_energy",   # or your desired TB scalar tag
    assume_tag_is_best=True,       # True: compare last point; False: compare global min
    make_plots=True,
    show_plots=True,
    # save_summary_csv=Path("./trained_vs_finetune_summary.csv"),
    # save_plot_path=Path("./trained_vs_finetune.png"),
)
```

```python
# ---------- TRAINED (CSV) vs TRAINED-FROM-SCRATCH (TensorBoard) ----------
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib as mpl
from typing import Dict, List, Tuple, Optional
from tensorboard.backend.event_processing import event_accumulator

# ====== you already have these in scope ======
# hex_ids: List[str]

# ---------- unified, export-friendly style (5 pt) ----------
mpl.rcParams.update({
    "font.family": "DejaVu Sans",
    "font.size": 5, "axes.titlesize": 5, "axes.labelsize": 5,
    "legend.fontsize": 5, "xtick.labelsize": 5, "ytick.labelsize": 5,
    "pdf.fonttype": 42, "ps.fonttype": 42, "pdf.use14corefonts": False, "svg.fonttype": "none",
    "axes.linewidth": 0.5, "lines.linewidth": 0.5,
    "legend.frameon": False, "savefig.dpi": 300, "figure.dpi": 100,
})
EDGE_COL, EDGE_W = "#58595B", 0.5
pt_to_in = lambda pt: pt / 72.0

def _style_axes(ax, grid_y=True):
    ax.tick_params(axis="y", which="major", width=EDGE_W, length=6, color=EDGE_COL)
    ax.tick_params(axis="y", which="minor", width=EDGE_W, length=3, color=EDGE_COL)
    ax.tick_params(axis="x", which="both",  width=EDGE_W, length=3, color=EDGE_COL)
    for s in ["left", "bottom"]:
        ax.spines[s].set_linewidth(EDGE_W)
        ax.spines[s].set_color(EDGE_COL)
    for s in ["top", "right"]:
        ax.spines[s].set_visible(False)
    if grid_y:
        ax.grid(True, axis="y", which="major", linewidth=0.3, alpha=0.3)

# ---- TB config for SCRATCH (TensorBoard) ----
SCRATCH_ROOT_TEMPLATE = (
    "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
    "manuscript/scratch_training/{HEX}/seed_1"
)
EVENT_GLOBS = ["**/events.out.tfevents.*", "**/*tfevents*"]
TB_TAG = "custom/best_energy"
ASSUME_TAG_IS_BEST = True  # True: compare final point; False: compare global min over series

# ---- Normalization ----
def _norm_hex(x: str) -> str:
    s = str(x).strip().lower()
    if not s.startswith("0x"): s = "0x" + s
    return "0x" + s[2:].upper()

# ---- TRAINED CSV discovery patterns ----
TRAINED_CSV_PATTERNS = [
    "trained_masked/*trained_episode_metrics.csv",
    "trained_masked/trajectories/*episode_metrics.csv",
    "**/*trained_episode_metrics.csv",
    "**/trajectories/*episode_metrics.csv",
    "trained_episode_metrics.csv",  # fallback at root
]

def _trained_root_for_hex(hex_id: str) -> Path:
    # Adjust to your actual trained directory layout if needed
    return Path(
        "/home/gridsan/spalacios/Designing complex biological circuits with deep neural networks/"
        f"manuscript/zero-shot/mp/5120_episodes/sf0/{_norm_hex(hex_id)}/seed_1"
    )

def _scratch_root_for_hex(hex_id: str) -> Path:
    return Path(SCRATCH_ROOT_TEMPLATE.format(HEX=_norm_hex(hex_id)))

def find_trained_csv(search_root: Path) -> Optional[Path]:
    root = Path(search_root)
    for pat in TRAINED_CSV_PATTERNS:
        hits = sorted(set(p for p in root.glob(pat)))
        if hits:
            return hits[0]
    return None

def get_trained_final_best(csv_path: Path) -> Optional[float]:
    """
    For trained per-episode metrics:
      - prefers 'best_energy_so_far'
      - else uses cummin('best_energy_in_episode')
      - returns the final best value (lower is better)
    """
    try:
        df = pd.read_csv(csv_path)
        if "episode" in df.columns:
            df["episode"] = pd.to_numeric(df["episode"], errors="coerce")
            df = df.dropna(subset=["episode"]).sort_values("episode", kind="mergesort")
        if "best_energy_so_far" in df.columns and df["best_energy_so_far"].notna().any():
            s = pd.to_numeric(df["best_energy_so_far"], errors="coerce").dropna()
            if len(s) > 0: return float(s.iloc[-1])
        if "best_energy_in_episode" in df.columns and df["best_energy_in_episode"].notna().any():
            e = pd.to_numeric(df["best_energy_in_episode"], errors="coerce").dropna()
            if len(e) > 0: return float(e.cummin().iloc[-1])
    except Exception as e:
        print(f"[WARN] Failed to read trained CSV {csv_path}: {e}")
    return None

# ---- Scratch (TensorBoard) helpers ----
def find_event_files(root: Path) -> List[Path]:
    files = []
    for pat in EVENT_GLOBS: files.extend(root.glob(pat))
    return sorted(set(files), key=lambda p: p.as_posix())

def load_tb_scalar_from_dir(run_dir: Path, tag: str) -> Tuple[np.ndarray, np.ndarray]:
    try:
        ea = event_accumulator.EventAccumulator(str(run_dir)); ea.Reload()
        if tag not in ea.Tags().get("scalars", []):
            return np.array([], dtype=int), np.array([], dtype=float)
        ev = ea.Scalars(tag)
        steps  = np.fromiter((e.step  for e in ev), dtype=int)
        values = np.fromiter((e.value for e in ev), dtype=float)
        return steps, values
    except Exception as e:
        print(f"[WARN] TB load failed in {run_dir}: {e}")
        return np.array([], dtype=int), np.array([], dtype=float)

def pick_best_from_runs(run_dirs: List[Path], tag: str, assume_best_series=True) -> Tuple[Optional[float], Optional[Path], np.ndarray, np.ndarray]:
    """Return (best_value, chosen_dir, steps, values) for the best run (lowest value)."""
    best_val, best_dir, best_steps, best_values = None, None, None, None
    for d in run_dirs:
        steps, vals = load_tb_scalar_from_dir(d, tag)
        if steps.size == 0: continue
        candidate = float(vals[-1]) if assume_best_series else float(np.nanmin(vals))
        if best_val is None or candidate < best_val:
            best_val, best_dir, best_steps, best_values = candidate, d, steps, vals
    return best_val, best_dir, best_steps, best_values

# ======== Plot helper (compact bar) ========
def _bar_figure_compact(categories: List[str], values: List[float], title: str) -> plt.Figure:
    fig, ax = plt.subplots(figsize=(pt_to_in(110), pt_to_in(90)), dpi=300)
    bars = ax.bar(categories, values, color=["#1976D2", "#8D6E63", "#9E9E9E"],
                  edgecolor=EDGE_COL, linewidth=EDGE_W)
    ax.set_ylim(0, 100)
    ax.set_ylabel("Percentage of circuits")
    ax.set_title(title)
    for i, v in enumerate(values):
        ax.text(i, min(100, v) + 0.5, f"{v:.1f}%", ha="center", va="bottom")
    _style_axes(ax, grid_y=True)
    fig.tight_layout()
    return fig

def _to_percentages(counts: Dict[str, int], denom: int) -> List[float]:
    if denom == 0: return [0.0, 0.0, 0.0]
    return [100.0*counts["Win"]/denom, 100.0*counts["Loss"]/denom, 100.0*counts["Tie"]/denom]

# ======== Core comparison: TRAINED vs SCRATCH ========
def compare_trained_to_scratch(
    hex_ids: List[str],
    tb_tag: str = TB_TAG,
    assume_tag_is_best: bool = ASSUME_TAG_IS_BEST,
    return_rows: bool = True,
):
    """
    Compare the final best energies between:
      - TRAINED: per-episode CSV (best_energy_so_far or cummin(best_energy_in_episode))
      - SCRATCH: TensorBoard scalar series (tag)
    Lower is better. 'Win' means TRAINED < SCRATCH.
    """
    outcomes = {"Win": 0, "Loss": 0, "Tie": 0}  # Win = TRAINED better
    missing, rows = [], []
    trained_hex, scratch_hex, tie_hex = [], [], []

    for hex_id in hex_ids:
        hex_norm = _norm_hex(hex_id)

        # TRAINED (CSV)
        tr_root = _trained_root_for_hex(hex_norm)
        tr_csv = find_trained_csv(tr_root)
        tr_final = get_trained_final_best(tr_csv) if tr_csv else None

        # SCRATCH (TensorBoard)
        sc_root = _scratch_root_for_hex(hex_norm)
        sc_events = find_event_files(sc_root)
        sc_run_dirs = sorted(set(p.parent for p in sc_events))
        sc_best, sc_dir, sc_steps, sc_vals = pick_best_from_runs(sc_run_dirs, tb_tag, assume_tag_is_best)

        if tr_final is None or sc_best is None:
            missing.append(hex_norm)
            continue

        if tr_final < sc_best:
            outcomes["Win"] += 1; verdict = "Win";  trained_hex.append(hex_norm)
        elif tr_final > sc_best:
            outcomes["Loss"] += 1; verdict = "Loss"; scratch_hex.append(hex_norm)
        else:
            outcomes["Tie"] += 1; verdict = "Tie";  tie_hex.append(hex_norm)

        if return_rows:
            rows.append({
                "hex_id": hex_norm,
                "trained_final": float(tr_final),
                "scratch_best": float(sc_best),
                "delta_trained_minus_scratch": float(tr_final - sc_best),
                "result": verdict,
                "trained_csv": str(tr_csv) if tr_csv else "",
                "scratch_chosen_run_dir": str(sc_dir) if sc_dir else "",
            })

    n_used = len(rows)
    if n_used == 0:
        win_pct = loss_pct = tie_pct = 0.0
    else:
        win_pct, loss_pct, tie_pct = _to_percentages(outcomes, n_used)

    # Pretty console summary
    print(f"[Trained vs Scratch] counts: {outcomes} out of N={n_used} (missing {len(missing)})")
    if trained_hex:
        print(f"  Trained better ({len(trained_hex)}): {', '.join(sorted(trained_hex))}")
    if scratch_hex:
        print(f"  Scratch better ({len(scratch_hex)}): {', '.join(sorted(scratch_hex))}")
    if tie_hex:
        print(f"  Ties ({len(tie_hex)}): {', '.join(sorted(tie_hex))}")
    if missing:
        print(f"  Missing ({len(missing)}): {', '.join(sorted(missing))}")

    per_hex_df = pd.DataFrame(rows)
    summary_df = pd.DataFrame([{
        "method": "trained_vs_scratch",
        "label": "Trained policy vs Trained-from-Scratch",
        "N_used": n_used,
        "Win": outcomes["Win"], "Loss": outcomes["Loss"], "Tie": outcomes["Tie"],
        "Win_%": win_pct, "Loss_%": loss_pct, "Tie_%": tie_pct,
        "N_missing": len(missing),
        "Missing_hex_ids": ",".join(sorted(missing)),
        "Trained_better_hex_ids": ",".join(sorted(trained_hex)),
        "Scratch_better_hex_ids": ",".join(sorted(scratch_hex)),
        "Tie_hex_ids": ",".join(sorted(tie_hex)),
        "tb_tag": tb_tag,
        "assume_tag_is_best": assume_tag_is_best,
    }])

    return summary_df, per_hex_df

# ======== Orchestrator (TRAINED vs SCRATCH) ========
def summarize_and_plot_trained_vs_scratch(
    hex_ids: List[str],
    tb_tag: str = TB_TAG,
    assume_tag_is_best: bool = ASSUME_TAG_IS_BEST,
    make_plots: bool = True,
    show_plots: bool = False,
    save_summary_csv: Optional[Path] = None,
    save_plot_path: Optional[Path] = None,
):
    """
    Returns (summary_df, per_hex_df, fig_or_None)
    """
    cats = ["Win", "Loss", "Tie"]
    summary, per_hex = compare_trained_to_scratch(hex_ids, tb_tag, assume_tag_is_best, return_rows=True)
    win_pct, loss_pct, tie_pct = float(summary["Win_%"].iloc[0]), float(summary["Loss_%"].iloc[0]), float(summary["Tie_%"].iloc[0])

    fig = None
    if make_plots:
        title = f"Trained vs Scratch (N={int(summary['N_used'].iloc[0])})  tag: {tb_tag}"
        fig = _bar_figure_compact(cats, [win_pct, loss_pct, tie_pct], title)
        if save_plot_path is not None:
            Path(save_plot_path).parent.mkdir(parents=True, exist_ok=True)
            fig.savefig(save_plot_path, bbox_inches="tight")
        if show_plots:
            plt.show()

    if save_summary_csv is not None:
        Path(save_summary_csv).parent.mkdir(parents=True, exist_ok=True)
        summary.to_csv(save_summary_csv, index=False)
        per_hex_out = save_summary_csv.with_name(save_summary_csv.stem + "_per_hex.csv")
        per_hex.to_csv(per_hex_out, index=False)

    return summary, per_hex, fig

# ======================== example usage ========================
# summary_df, per_hex_df, fig = summarize_and_plot_trained_vs_scratch(
#     hex_ids,
#     tb_tag="custom/best_energy",   # or your desired TB scalar tag
#     assume_tag_is_best=True,       # True: compare last point; False: compare global min
#     make_plots=True,
#     show_plots=True,
#     # save_summary_csv=Path("./trained_vs_scratch_summary.csv"),
#     # save_plot_path=Path("./trained_vs_scratch.png"),
# )

# ======================== example usage (active) ========================
summary_df, per_hex_df, fig = summarize_and_plot_trained_vs_scratch(
    hex_ids,
    tb_tag="custom/best_energy",   # or your desired TB scalar tag
    assume_tag_is_best=True,       # True: compare last point; False: compare global min
    make_plots=True,
    show_plots=True,
    # save_summary_csv=Path("./trained_vs_scratch_summary.csv"),
    # save_plot_path=Path("./trained_vs_scratch.png"),
)
```
