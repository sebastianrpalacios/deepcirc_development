```python
def load_graph_pickle(filename):
    """
    Load a graph from a pickle file and convert back to NetworkX format.
    
    Args:
        filename (str): Pickle file to load
        
    Returns:
        nx.DiGraph: Reconstructed graph
    """
    # Load the list from pickle
    with open(filename, 'rb') as f:
        graph_list = pickle.load(f)
    
    # Extract components
    num_nodes, edges, node_attrs = graph_list
    
    # Create new graph
    G = nx.DiGraph()
    
    # Add nodes with attributes
    for node, attr in node_attrs.items():
        if attr is not None:
            G.add_node(node, type=attr)
        else:
            G.add_node(node)
    
    # Add edges
    G.add_edges_from(edges)
    
    return G
```

```python


import os, re, pickle, networkx as nx          # networkx needed by loader
from tqdm.auto import tqdm                     # Jupyter-friendly progress bar

# ---------------------------------------------------------------------
# 0.  Configuration
# ---------------------------------------------------------------------
NIG_DIR = ("/home/gridsan/spalacios/Designing complex biological circuits "
           "with deep neural networks/dgd/data/NIGs_4_inputs/")
hex_re = re.compile(r"0x([0-9a-fA-F]+)")       # leading 0xHEX pattern


# ---------------------------------------------------------------------
# 3.  Verification loop
# ---------------------------------------------------------------------
ok, mismatches, errors = [], [], []
pkl_files = [f for f in os.listdir(NIG_DIR) if f.endswith(".pkl")]

for fname in tqdm(pkl_files, desc="Verifying graphs"):
    m = hex_re.match(fname)
    if not m:
        tqdm.write(f"skipped (no 0xHEX_ prefix): {fname}")
        continue

    n_file = int(m.group(1), 16)                 # int from filename hex
    fpath  = os.path.join(NIG_DIR, fname)

    try:
        g = load_graph_pickle(fpath)             # <<— uses your loader
        inputs, n_calc = _compute_truth_key(g)

        if n_calc == n_file:
            ok.append(fname)
        else:
            mismatches.append((fname, inputs, n_file, n_calc))

    except Exception as e:
        errors.append((fname, str(e)))
        tqdm.write(f"{fname}: {e}")

# ---------------------------------------------------------------------
# 4.  Summary
# ---------------------------------------------------------------------
print("\n=== Verification summary ===")
print(f"✓ {len(ok)} graphs correct")
print(f"✗ {len(mismatches)} filename ≠ truth-table")
print(f"⚠️ {len(errors)} failed to load / evaluate")

if mismatches:
    print("\nMISMATCHES (file_hex ≠ computed_hex):")
    for fname, inputs, n_file, n_calc in mismatches:
        print(f"  {fname}: inputs={inputs}, 0x{n_file:04x} ≠ 0x{n_calc:04x}")

if errors:
    print("\nERRORS:")
    for fname, msg in errors:
        print(f"  {fname}: {msg}")
```
