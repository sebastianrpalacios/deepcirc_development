```python
"""
Check for two requirements. 

1. It contains: Total number of inputs: 3
2. That there is a Score: in the file. 
    
If those requirenemnts are not met, then skip the file. 

""" 

import os
import shutil

def process_log_files(source_folder, destination_folder):
    for root, dirs, files in os.walk(source_folder):
        for file in files:
            if file == "log.log":
                source_file_path = os.path.join(root, file)
                
                # Read the file and check for requirements
                with open(source_file_path, 'r') as f:
                    content = f.read()
                
                has_total_inputs = 'Total number of inputs: 3' in content
                has_score = 'Score:' in content

                # Check the requirements
                if has_total_inputs and has_score:
                    subfolder_name = os.path.basename(root)
                    new_file_name = "{}_log.log".format(subfolder_name)
                    destination_file_path = os.path.join(destination_folder, new_file_name)
                    
                    # Duplicate and rename the file
                    shutil.copy2(source_file_path, destination_file_path)
                    print("Copied {} to {}".format(source_file_path, destination_file_path))
                else:
                    print("Skipped {}: Requirements not met".format(source_file_path))

# Example usage
source_folder = 'output_run_postech_1'  # Replace with the path to your source folder
destination_folder = 'output_run_postech_1_logfiles'  # Replace with the path to your destination folder

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Process the log files
process_log_files(source_folder, destination_folder)
```

```python
import os
import shutil

def process_json_files(source_folder, destination_folder):
    for root, dirs, files in os.walk(source_folder):
        for file in files:
            if file.endswith("outputNetlist.json"):
                source_file_path = os.path.join(root, file)
                destination_file_path = os.path.join(destination_folder, file)
                
                # Check if the file already exists in the destination folder to avoid overwriting
                if os.path.exists(destination_file_path):
                    print("File {} already exists. Skipping...".format(destination_file_path))
                else:
                    # Duplicate the file
                    shutil.copy2(source_file_path, destination_file_path)
                    print("Copied {} to {}".format(source_file_path, destination_file_path))

# Example usage
source_folder = 'output_run_postech_1'  # Replace with the path to your source folder
destination_folder = 'output_run_postech_1_jsonfiles'  # Replace with the path to your destination folder

# Create the destination folder if it doesn't exist
os.makedirs(destination_folder, exist_ok=True)

# Process the JSON files
process_json_files(source_folder, destination_folder)

```
